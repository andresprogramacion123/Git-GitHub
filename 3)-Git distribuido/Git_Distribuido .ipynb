## Git distribuido

### Flujos de trabajo distribuidos

Ahora que tiene un repositorio Git remoto configurado como punto focal para que todos los desarrolladores compartan su código, y está familiarizado con los comandos básicos de Git en un flujo de trabajo local, verá cómo utilizar algunos de los flujos de trabajo distribuido que te ofrece Git.

En este capítulo, verá cómo trabajar con Git en un entorno distribuido como colaborador e integrador. Es decir, aprenderá cómo contribuir con código con éxito a un proyecto y hacerlo lo más fácil posible para usted y el encargado del mantenimiento del proyecto, y también cómo mantener un proyecto con éxito con la contribución de varios desarrolladores.

A diferencia de los sistemas de control de versiones centralizados (CVCS), la naturaleza distribuida de Git le permite ser mucho más flexible en la forma en que los desarrolladores colaboran en los proyectos. En los sistemas centralizados, cada desarrollador es un nodo que trabaja más o menos por igual con un hub central. En Git, sin embargo, cada desarrollador es potencialmente tanto un nodo como un concentrador; es decir, cada desarrollador puede contribuir con código a otros repositorios y mantener un repositorio público en el que otros pueden basar su trabajo y en el que pueden contribuir. Esto presenta una amplia gama de posibilidades de flujo de trabajo para su proyecto y / o su equipo, por lo que cubriremos algunos paradigmas comunes que aprovechan esta flexibilidad. Repasaremos las fortalezas y posibles debilidades de cada diseño; puede elegir uno solo para usar, o puede mezclar y combinar características de cada uno.

### Flujo de trabajo centralizado

En los sistemas centralizados, generalmente existe un único modelo de colaboración: el flujo de trabajo centralizado. Un centro o repositorio central puede aceptar código y todos sincronizan su trabajo con él. Varios desarrolladores son nodos, consumidores de ese centro, y se sincronizan con esa ubicación centralizada.

![](Images/img1.png)

Esto significa que si dos desarrolladores clonan desde el concentrador y ambos realizan cambios, el primer desarrollador que haga una copia de seguridad de los cambios puede hacerlo sin problemas. El segundo desarrollador debe fusionarse en el trabajo del primero antes de subir los cambios, para no sobrescribir los cambios del primer desarrollador. Este concepto es tan cierto en Git como en Subversion (o cualquier CVCS), y este modelo funciona perfectamente bien en Git.

Si ya se siente cómodo con un flujo de trabajo centralizado en su empresa o equipo, puede continuar usando ese flujo de trabajo fácilmente con Git. Simplemente configure un único repositorio y dé a todos los miembros de su equipo acceso push; Git no permitirá que los usuarios se sobrescriban entre sí.

Digamos que John y Jessica empiezan a trabajar al mismo tiempo. John termina su cambio y lo envía al servidor. Entonces Jessica intenta impulsar sus cambios, pero el servidor los rechaza. Se le dice que está tratando de impulsar cambios que no sean de avance rápido y que no podrá hacerlo hasta que busque y se fusione. Este flujo de trabajo es atractivo para mucha gente porque es un paradigma con el que muchos están familiarizados y se sienten cómodos.

Esto tampoco se limita a equipos pequeños. Con el modelo de ramificación de Git, es posible que cientos de desarrolladores trabajen con éxito en un solo proyecto a través de docenas de ramificaciones simultáneamente.

### Flujo de trabajo de Integration-Manager

Debido a que Git le permite tener múltiples repositorios remotos, es posible tener un flujo de trabajo en el que cada desarrollador tenga acceso de escritura a su propio repositorio público y acceso de lectura al de todos los demás. Este escenario a menudo incluye un repositorio canónico que representa el proyecto "oficial". Para contribuir a ese proyecto, crea su propio clon público del proyecto y envía sus cambios. Luego, puede enviar una solicitud al mantenedor del proyecto principal para que realice los cambios. El mantenedor puede luego agregar su repositorio como un control remoto, probar sus cambios localmente, fusionarlos en su rama y regresar a su repositorio. El proceso funciona de la siguiente manera (consulte Flujo de trabajo del administrador de integración):

1. El mantenedor del proyecto empuja (push) a su repositorio público.

2. Un colaborador clona ese repositorio y realiza cambios.

3. El colaborador empuja a su propia copia pública.

4. El colaborador envía al mantenedor un correo electrónico pidiéndole que realice los cambios.

5. El mantenedor agrega el repositorio del colaborador como remoto y lo fusiona localmente.

6. El mantenedor envía los cambios combinados al repositorio principal.

![](Images/img2.png)

Este es un flujo de trabajo muy común con herramientas basadas en concentradores como GitHub o GitLab, donde es fácil bifurcar un proyecto e introducir los cambios en tu bifurcación para que todos los vean. Una de las principales ventajas de este enfoque es que puede seguir trabajando y el responsable del mantenimiento del repositorio principal puede introducir sus cambios en cualquier momento. Los colaboradores no tienen que esperar a que el proyecto incorpore sus cambios; cada parte puede trabajar a su propio ritmo.

### Flujo de trabajo de dictador y tenientes

Esta es una variante de un flujo de trabajo de repositorios múltiples. Generalmente se usa en proyectos grandes con cientos de colaboradores; un ejemplo famoso es el kernel de Linux. Varios administradores de integración están a cargo de ciertas partes del repositorio; se llaman tenientes. Todos los tenientes tienen un gerente de integración conocido como el dictador benevolente. El dictador benevolente empuja (push) desde su directorio a un repositorio de referencia del que todos los colaboradores deben extraer. El proceso funciona así (consulte Flujo de trabajo de dictador benevolente):

1. Los desarrolladores habituales trabajan en su topic branch y vuelven a combinar su trabajo en la parte superior master. La rama master es la del repositorio de referencia al que empuja (push) el dictador.

2. Los tenientes fusionan (merge) las ramas temáticas de los desarrolladores en su rama master.

3. El dictador fusiona (merge) las ramas master de los lugar tenientes en la rama master del dictador.

4. Finalmente, el dictador empuja (push) esa rama master al repositorio de referencia para que los otros desarrolladores puedan volver a basarse en ella.

![](Images/img3.png)

Este tipo de flujo de trabajo no es común, pero puede ser útil en proyectos muy grandes o en entornos altamente jerárquicos. Permite al líder del proyecto (el dictador) delegar gran parte del trabajo y recopilar grandes subconjuntos de código en múltiples puntos antes de integrarlos.

### Contribuir a un proyecto

La principal dificultad para describir cómo contribuir a un proyecto son las numerosas variaciones sobre cómo hacerlo. Debido a que Git es muy flexible, las personas pueden trabajar y trabajan juntas de muchas maneras, y es problemático describir cómo debes contribuir: cada proyecto es un poco diferente. Algunas de las variables involucradas son el recuento de contribuyentes activos, el flujo de trabajo elegido, su acceso de compromiso y posiblemente el método de contribución externa.

La primera variable es el recuento de contribuyentes activos: ¿cuántos usuarios están contribuyendo activamente con el código a este proyecto y con qué frecuencia? En muchos casos, tendrá dos o tres desarrolladores con algunas confirmaciones al día, o posiblemente menos para proyectos algo inactivos. Para empresas o proyectos más grandes, la cantidad de desarrolladores podría ser de miles, con cientos o miles de confirmaciones cada día. Esto es importante porque con más y más desarrolladores, se encuentra con más problemas para asegurarse de que su código se aplique de manera limpia o se pueda fusionar fácilmente. Los cambios que envíe pueden volverse obsoletos o muy dañados por el trabajo que se fusionó mientras estaba trabajando o mientras sus cambios esperaban ser aprobados o aplicados. ¿Cómo puede mantener su código constantemente actualizado y sus confirmaciones válidas?

La siguiente variable es el flujo de trabajo en uso para el proyecto. ¿Está centralizado y cada desarrollador tiene el mismo acceso de escritura a la línea de código principal? ¿El proyecto tiene un responsable de mantenimiento o de integración que verifica todos los parches? ¿Todos los parches se revisan y aprueban por pares? ¿Estás involucrado en ese proceso? ¿Existe un sistema de tenientes y primero debe enviarles su trabajo?

La siguiente variable es su acceso de confirmación. El flujo de trabajo necesario para contribuir a un proyecto es muy diferente si tiene acceso de escritura al proyecto que si no lo tiene. Si no tiene acceso de escritura, ¿cómo prefiere el proyecto aceptar el trabajo contribuido? ¿Tiene siquiera una política? ¿Cuánto trabajo estás contribuyendo a la vez? ¿Con qué frecuencia contribuyes?

Todas estas preguntas pueden afectar su contribución efectiva a un proyecto y qué flujos de trabajo son preferidos o disponibles para usted. Cubriremos aspectos de cada uno de estos en una serie de casos de uso, pasando de simples a más complejos; debería poder construir los flujos de trabajo específicos que necesita en la práctica a partir de estos ejemplos.

### Pautas de compromiso

Antes de comenzar a ver los casos de uso específicos, aquí hay una nota rápida sobre los mensajes de confirmación. Tener una buena guía para crear confirmaciones y ceñirse a ella hace que trabajar con Git y colaborar con otros sea mucho más fácil. El proyecto Git proporciona un documento que presenta una serie de buenos consejos para crear confirmaciones a partir de las cuales enviar parches; puede leerlo en el código fuente de Git en el archivo ```Documentation/SubmittingPatches```.

Primero, sus envíos no deben contener errores de espacio en blanco. Git proporciona una manera fácil de verificar esto: antes de confirmar, ejecutar ```git diff --check```, lo que identifica posibles errores de espacios en blanco y los enumera por usted.

![](Images/img4.png)

Si ejecuta ese comando antes de confirmar, puede saber si está a punto de cometer problemas de espacios en blanco que pueden molestar a otros desarrolladores.

A continuación, intente hacer que cada confirmación sea un conjunto de cambios lógicamente separado. Si puede, intente hacer que sus cambios sean digeribles: no codifique durante todo un fin de semana en cinco problemas diferentes y luego envíelos todos como una confirmación masiva el lunes. Incluso si no te comprometes durante el fin de semana, usa el área de preparación el lunes para dividir tu trabajo en al menos un compromiso por problema, con un mensaje útil por compromiso. Si algunos de los cambios modifican el mismo archivo, intente utilizarlo ```git add --patch``` para organizar archivos parcialmente (se explica en detalle en la puesta en escena interactiva).El snapshot del proyecto en la punta de la rama es idéntica ya sea que realice una confirmación o cinco, siempre que todos los cambios se agreguen en algún momento, así que trate de facilitar las cosas a sus compañeros desarrolladores cuando tengan que revisar sus cambios.

Este enfoque también facilita la extracción (pull) o reversión de uno de los conjuntos de cambios si es necesario más adelante. Reescritura del historial describe una serie de trucos útiles de Git para reescribir el historial y organizar archivos de forma interactiva: use estas herramientas para ayudar a crear un historial limpio y comprensible antes de enviar el trabajo a otra persona.

Lo último a tener en cuenta es el mensaje de confirmación. Adquirir el hábito de crear mensajes de confirmación de calidad facilita mucho el uso y la colaboración con Git. Como regla general, sus mensajes deben comenzar con una sola línea que no tenga más de 50 caracteres y que describa el conjunto de cambios de manera concisa, seguida de una línea en blanco, seguida de una explicación más detallada. El proyecto Git requiere que la explicación más detallada incluya su motivación para el cambio y contraste su implementación con el comportamiento anterior; esta es una buena guía a seguir. Escriba su mensaje de confirmación en imperativo: "Corregir error" y no "Corregi errores" o "Correges errores". Aquí hay una plantilla que puede seguir, que hemos adaptado a la ligera de una escrita originalmente por Tim Pope:

```
Capitalized, short (50 chars or less) summary

More detailed explanatory text, if necessary. Wrap it to about 72 characters or so.  In some contexts, the first line is treated as the subject of an email and the rest of the text as the body.  The blank line separating the summary from the body is critical (unless you omit the body entirely); tools like rebase will confuse you if you run the
two together.

Write your commit message in the imperative: "Fix bug" and not "Fixed bug" or "Fixes bug."  This convention matches up with commit messages generated by commands like git merge and git revert.

Further paragraphs come after blank lines.

- Bullet points are okay, too

- Typically a hyphen or asterisk is used for the bullet, followed by a single space, with blank lines in between, but conventions vary here

- Use a hanging indent
```
Si todos sus mensajes de confirmación siguen este modelo, las cosas serán mucho más fáciles para usted y los desarrolladores con los que colabora. El proyecto Git tiene mensajes de confirmación bien formateados; intente ejecutarlo ```git log --no-merges``` allí para ver cómo se ve un historial de confirmación de proyecto bien formateado.

**Nota**

Haz lo que decimos, no lo que hacemos.

En aras de la brevedad, muchos de los ejemplos de este libro no tienen mensajes de confirmación con un formato agradable como este; en su lugar, simplemente usamos la opción ```-m``` para ```git commit```.

En resumen, haz lo que decimos, no lo que hacemos.

### Pequeño equipo privado

La configuración más simple que probablemente encontrará es un proyecto privado con uno o dos desarrolladores más. “Privado”, en este contexto, significa código cerrado, no accesible al mundo exterior. Usted y los otros desarrolladores tienen acceso ```push``` al repositorio.

En este entorno, puede seguir un flujo de trabajo similar al que podría hacer cuando usa Subversion u otro sistema centralizado. Aún obtiene las ventajas de cosas como la confirmación fuera de línea y la ramificación y fusión mucho más simple, pero el flujo de trabajo puede ser muy similar; la principal diferencia es que las fusiones ocurren en el lado del cliente en lugar de en el servidor en el momento de la confirmación. Veamos cómo se vería cuando dos desarrolladores comenzaran a trabajar juntos con un repositorio compartido. El primer desarrollador, John, clona el repositorio, realiza un cambio y confirma localmente. Los mensajes de protocolo se han reemplazado con ... en estos ejemplos para acortarlos un poco.

```
# John's Machine
$ git clone john@githost:simplegit.git
Cloning into 'simplegit'...
...
$ cd simplegit/
$ vim lib/simplegit.rb
$ git commit -am 'Remove invalid default value'
[master 738ee87] Remove invalid default value
 1 files changed, 1 insertions(+), 1 deletions(-)
```

La segunda desarrolladora, Jessica, hace lo mismo: clona el repositorio y realiza un cambio:

```
# Jessica's Machine
$ git clone jessica@githost:simplegit.git
Cloning into 'simplegit'...
...
$ cd simplegit/
$ vim TODO
$ git commit -am 'Add reset task'
[master fbff5bc] Add reset task
 1 files changed, 1 insertions(+), 0 deletions(-)
```

Ahora, Jessica envía su trabajo al servidor, que funciona bien:

```
# Jessica's Machine
$ git push origin master
...
To jessica@githost:simplegit.git
   1edee6b..fbff5bc  master -> master
```

La última línea de la salida anterior muestra un mensaje útil de retorno de la operación ```push```. El formato básico es ```<oldref>..<newref> fromref → toref```, donde ```oldref``` significa la referencia anterior, ```newref``` significa la nueva referencia, ```fromref``` es el nombre de la referencia local que se está haciendo push y ```toref``` es el nombre de la referencia remota que se está actualizando. Verá un resultado similar como este a continuación en las discusiones, por lo que tener una idea básica del significado ayudará a comprender los diversos estados de los repositorios. Hay más detalles disponibles en la documentación de ```git-push```.

Continuando con este ejemplo, poco después, John realiza algunos cambios, los confirma en su repositorio local e intenta enviarlos al mismo servidor:

```
# John's Machine
$ git push origin master
To john@githost:simplegit.git
 ! [rejected]        master -> master (non-fast forward)
error: failed to push some refs to 'john@githost:simplegit.git'
```

En este caso, el push de John falla debido al push anterior de Jessica de sus cambios. Es especialmente importante entender esto si está acostumbrado a Subversion, porque notará que los dos desarrolladores no editaron el mismo archivo. Aunque Subversion automáticamente hace tal fusión en el servidor si se editan diferentes archivos, con Git, primero debe fusionar las confirmaciones localmente. En otras palabras, John primero debe buscar los cambios ascendentes de Jessica y fusionarlos en su repositorio local antes de que se le permita enviar.

Como primer paso, John busca el trabajo de Jessica (esto solo busca el trabajo anterior de Jessica, todavía no lo fusiona con el trabajo de John):

```
$ git fetch origin
...
From john@githost:simplegit
 + 049d078...fbff5bc master     -> origin/master
```

En este punto, el repositorio local de John se parece a esto:

![](Images/img5.png)

Ahora John puede fusionar el trabajo de Jessica que obtuvo en su propio trabajo local:

```
$ git merge origin/master
Merge made by the 'recursive' strategy.
 TODO |    1 +
 1 files changed, 1 insertions(+), 0 deletions(-)
```

Siempre que la fusión local se realice sin problemas, el historial actualizado de John ahora se verá así:

![](Images/img6.png)

En este punto, es posible que John quiera probar este nuevo código para asegurarse de que ninguno de los trabajos de Jessica afecte a ninguno de los suyos y, siempre que todo parezca estar bien, finalmente puede enviar el nuevo trabajo combinado al servidor:

```
$ git push origin master
...
To john@githost:simplegit.git
   fbff5bc..72bbc59  master -> master
```

Al final, el historial de confirmaciones de John se verá así:

![](Images/img7.png)

Mientras tanto, Jessica ha creado una nueva topic branch llamada ```issue54``` y ha realizado tres confirmaciones para esa rama. Aún no ha obtenido los cambios de John, por lo que su historial de confirmaciones se ve así:

![](Images/img8.png)

De repente, Jessica se entera de que John ha enviado un nuevo trabajo al servidor y quiere echarle un vistazo, para poder obtener todo el contenido nuevo del servidor que aún no tiene con:

```
# Jessica's Machine
$ git fetch origin
...
From jessica@githost:simplegit
   fbff5bc..72bbc59  master     -> origin/master
```

Eso reduce el trabajo que John ha impulsado mientras tanto. La historia de Jessica ahora se ve así:

![](Images/img9.png)

Jessica cree que su rama temática está lista, pero quiere saber qué parte del trabajo buscado de John tiene que fusionar con su trabajo para poder seguir adelante. Ella corre ```git log``` para averiguar:

```
$ git log --no-merges issue54..origin/master
commit 738ee872852dfaa9d6634e0dea7a324040193016
Author: John Smith <jsmith@example.com>
Date:   Fri May 29 16:01:27 2009 -0700

   Remove invalid default value
```

La sintaxis ```issue54..origin/master``` es un filtro de registro que le pide a Git que muestre solo aquellas confirmaciones que están en la última rama (en este caso ```origin/master```) que no están en la primera rama (en este caso ```issue54```). Repasaremos esta sintaxis en detalle en Commit Ranges.

A partir de la salida anterior, podemos ver que hay una única confirmación que John ha hecho de que Jessica no se ha fusionado con su trabajo local. Si se fusiona ```origin/master```, esa es la única confirmación que modificará su trabajo local.

Ahora, Jessica puede fusionar su trabajo temático en su rama master, fusionar el trabajo de John (```origin/master```) en su rama master y luego volver al servidor.

Primero (habiendo comprometido todo el trabajo en su rama temática ```issue54```), Jessica vuelve a su rama master en preparación para integrar todo este trabajo:

```
$ git checkout master
Switched to branch 'master'
Your branch is behind 'origin/master' by 2 commits, and can be fast-forwarded.
```

Jessica puede fusionar uno ```origin/master``` o el ```issue54``` primero: ambos están en sentido ascendente, por lo que el orden no importa. El snapshot final debe ser idéntica sin importar el orden que elija; solo la historia será diferente. Ella elige fusionar la rama ```issue54``` primero:

```
$ git merge issue54
Updating fbff5bc..4af4298
Fast forward
 README           |    1 +
 lib/simplegit.rb |    6 +++++-
 2 files changed, 6 insertions(+), 1 deletions(-)
```

No ocurren problemas; como puede ver, fue una simple combinación de avance rápido. Jessica ahora completa el proceso de fusión local fusionando el trabajo obtenido anteriormente de John que se encuentra en la branch ```origin/master```:

```
$ git merge origin/master
Auto-merging lib/simplegit.rb
Merge made by the 'recursive' strategy.
 lib/simplegit.rb |    2 +-
 1 files changed, 1 insertions(+), 1 deletions(-)
``` 

Todo se fusiona limpiamente, y la historia de Jessica ahora se ve así:

![](Images/img10.png)

Ahora ```origin/master``` se puede acceder desde la rama master de Jessica , por lo que debería poder presionar (push) con éxito (suponiendo que John no haya impulsado aún más cambios mientras tanto):

```
$ git push origin master
...
To jessica@githost:simplegit.git
   72bbc59..8059c15  master -> master
```

Cada desarrollador se ha comprometido varias veces y ha fusionado el trabajo de los demás con éxito.

![](Images/img11.png)

Ese es uno de los flujos de trabajo más simples. Trabaja por un tiempo (generalmente en una topic branch) y fusiona ese trabajo en su rama master cuando está listo para integrarse. Cuando desea compartir ese trabajo, busca y fusiona su master desde ```origin/master``` si ha cambiado, y finalmente empuja (push) a la rama master en el servidor. La secuencia general es algo como esto:

![](Images/img12.png)

### Equipo administrado privado

En este próximo escenario, verá los roles de los contribuyentes en un grupo privado más grande. Aprenderá a trabajar en un entorno donde pequeños grupos colaboran en funciones, después de lo cual esas contribuciones basadas en equipo son integradas por otra parte.

Digamos que John y Jessica están trabajando juntos en una característica (llamémosla "característica A"), mientras que Jessica y un tercer desarrollador, Josie, están trabajando en una segunda (digamos "característica B"). En este caso, la empresa está utilizando un tipo de flujo de trabajo de administrador de integración donde el trabajo de los grupos individuales está integrado solo por ciertos ingenieros, y la rama master del repositorio principal solo puede ser actualizada por esos ingenieros. En este escenario, todo el trabajo se realiza en ramas basadas en equipos y los integradores lo reúnen más tarde.

Sigamos el flujo de trabajo de Jessica mientras trabaja en sus dos funciones, colaborando en paralelo con dos desarrolladores diferentes en este entorno. Suponiendo que ya ha clonado su repositorio, decide trabajar primero "característica A". Ella crea una nueva rama para la función y trabaja en ella allí:

```
# Jessica's Machine
$ git checkout -b featureA
Switched to a new branch 'featureA'
$ vim lib/simplegit.rb
$ git commit -am 'Add limit to log function'
[featureA 3300904] Add limit to log function
 1 files changed, 1 insertions(+), 1 deletions(-)
```

En este punto, necesita compartir su trabajo con John, por lo que envía sus confirmaciones "característica A" de branch al servidor. Jessica no tiene acceso push a la rama master, solo los integradores lo tienen, por lo que tiene que pasar a otra rama para colaborar con John:

```
$ git push -u origin featureA
...
To jessica@githost:simplegit.git
 * [new branch]      featureA -> featureA
```

Jessica le envía un correo electrónico a John para decirle que ha enviado algunos trabajos a una rama con nombre featureA y que él puede verlo ahora. Mientras espera los comentarios de John, Jessica decide empezar a trabajar featureB con Josie. Para empezar, inicia una nueva rama de funciones, basándola en la rama master del servidor:

```
# Jessica's Machine
$ git fetch origin
$ git checkout -b featureB origin/master
Switched to a new branch 'featureB'
```

Ahora, Jessica realiza un par de confirmaciones en la rama featureB:

```
$ vim lib/simplegit.rb
$ git commit -am 'Make ls-tree function recursive'
[featureB e5b0fdc] Make ls-tree function recursive
 1 files changed, 1 insertions(+), 1 deletions(-)
$ vim lib/simplegit.rb
$ git commit -am 'Add ls-files'
[featureB 8512791] Add ls-files
 1 files changed, 5 insertions(+), 0 deletions(-)
```

El repositorio de Jessica ahora se ve así:

![](Images/img13.png)

Ella está lista para impulsar **(push)** su trabajo, pero recibe un correo electrónico de Josie que dice que una rama con un trabajo inicial de "featureB" ya se envió al servidor como rama "featureBee". Jessica necesita fusionar esos cambios con los suyos antes de poder enviar su trabajo al servidor. Jessica primero busca los cambios de Josie con git fetch:

```
$ git fetch origin
...
From jessica@githost:simplegit
 * [new branch]      featureBee -> origin/featureBee
```

Suponiendo que Jessica todavía está en su rama featureB cerrada, ahora puede fusionar el trabajo de Josie en esa rama con ```git merge```:

```
$ git merge origin/featureBee
Auto-merging lib/simplegit.rb
Merge made by the 'recursive' strategy.
 lib/simplegit.rb |    4 ++++
 1 files changed, 4 insertions(+), 0 deletions(-)
```

En este punto, Jessica quiere enviar todo este trabajo de "featureB" combinado al servidor, pero no quiere simplemente empujar su propia rama featureB. Por el contrario, dado que Josie ya ha comenzado una rama  "featureBee" aguas arriba , Jessica quiere avanzar a esa rama, lo que hace con:

```
$ git push -u origin featureB:featureBee
...
To jessica@githost:simplegit.git
   fba9af8..cd685d1  featureB -> featureBee
``` 

Esto se llama **refspec**. Consulte The Refspec para obtener una descripción más detallada de las especificaciones de referencia de Git y las diferentes cosas que puede hacer con ellas. También observe la bandera ```-u```; esto es la abreviatura de ```--set-upstream```, que configura las ramas para empujar y tirar más fácilmente más tarde.

De repente, Jessica recibe un correo electrónico de John, quien le dice que ha impulsado algunos cambios en la rama featureA  en la que están colaborando, y le pide a Jessica que les eche un vistazo. Una vez más, Jessica ejecuta un sencillo ```git fetch``` para buscar todo el contenido nuevo del servidor, incluido (por supuesto) el último trabajo de John:

```
$ git fetch origin
...
From jessica@githost:simplegit
   3300904..aad881d  featureA   -> origin/featureA
```

Jessica puede mostrar el registro del nuevo trabajo de John comparando el contenido de la rama featureA recién obtenida con su copia local de la misma rama:

```
$ git log featureA..origin/featureA
commit aad881d154acdaeb2b6b18ea0e827ed8a6d671e6
Author: John Smith <jsmith@example.com>
Date:   Fri May 29 19:57:33 2009 -0700

    Increase log output to 30 from 25
```

Si a Jessica le gusta lo que ve, puede fusionar el nuevo trabajo de John en su rama local featureA  con:

```
$ git checkout featureA
Switched to branch 'featureA'
$ git merge origin/featureA
Updating 3300904..aad881d
Fast forward
 lib/simplegit.rb |   10 +++++++++-
1 files changed, 9 insertions(+), 1 deletions(-)
```

Finalmente, Jessica podría querer hacer un par de cambios menores en todo el contenido combinado, por lo que es libre de realizar esos cambios, enviarlos a su rama local featureA  y enviar el resultado final al servidor:

```
$ git commit -am 'Add small tweak to merged content'
[featureA 774b3ed] Add small tweak to merged content
 1 files changed, 1 insertions(+), 1 deletions(-)
$ git push
...
To jessica@githost:simplegit.git
   3300904..774b3ed  featureA -> featureA
```

### Proyecto público bifurcado (fork)

Contribuir a proyectos públicos es un poco diferente. Debido a que no tiene los permisos para actualizar directamente las ramas en el proyecto, debe llevar el trabajo a los encargados de mantenimiento de alguna otra manera. Este primer ejemplo describe la contribución mediante bifurcación en Git hosts que admiten bifurcación fácil. Muchos sitios de alojamiento admiten esto (incluidos GitHub, BitBucket, repo.or.cz y otros), y muchos encargados de mantenimiento de proyectos esperan este estilo de contribución. La siguiente sección trata sobre proyectos que prefieren aceptar parches aportados por correo electrónico.

Primero, probablemente querrá clonar el repositorio principal, crear una topic branch para el parche o la serie de parches que planea contribuir y hacer su trabajo allí. La secuencia se ve básicamente así:

```
$ git clone <url>
$ cd project
$ git checkout -b featureA
  ... work ...
$ git commit
  ... work ...
$ git commit
```

**Nota**

Es posible que desee utilizar ```rebase -i``` para reducir su trabajo a una única confirmación, o reorganizar el trabajo en las confirmaciones para que el parche sea más fácil de revisar para el mantenedor; consulte Reescritura del historial para obtener más información sobre el reajuste interactivo.

Cuando el trabajo de la rama esté terminado y esté listo para contribuir con los mantenedores, vaya a la página del proyecto original y haga clic en el botón **"Bifurcación" (fork)**, creando su propia bifurcación del proyecto que se puede escribir. Luego, debe agregar esta URL del repositorio como un nuevo control remoto de su repositorio local; en este ejemplo, llamémoslo ```myfork```:

```
$ git remote add myfork <url>
```

Luego, debe enviar su nuevo trabajo a este repositorio. Es más fácil enviar la rama temática en la que está trabajando a su repositorio bifurcado, en lugar de fusionar ese trabajo en su rama master y empujarlo. La razón es que si su trabajo no se acepta o se selecciona con precisión, no tiene que rebobinar su rama ```master``` (la operación cherry-pick de Git se trata con más detalle en Rebasing y Cherry-Picking Workflows ). Si los mantenedores ```merge```, ```rebase``` o ```cherry-pick``` su trabajo, eventualmente lo recuperarán extrayéndolo de su repositorio de todos modos.

En cualquier caso, puede impulsar (push) su trabajo con:

```
$ git push -u myfork featureA
```

Una vez que su trabajo ha sido enviado a su bifurcación del repositorio, debe notificar a los encargados del mantenimiento del proyecto original que tiene trabajo que le gustaría que fusionaran. Esto a menudo se llama una solicitud de **extracción(Pull request)**, y normalmente genera dicha solicitud a través del sitio web (GitHub tiene su propio mecanismo de "Solicitud de extracción " que revisaremos en GitHub) o puede ejecutar el comando ```git request-pull``` y enviar por correo electrónico la salida posterior a el mantenedor del proyecto manualmente.

El comando ```git request-pull``` toma la rama base en la que desea que se extraiga la rama de su tema y la URL del repositorio de Git de la que desea que extraigan, y produce un resumen de todos los cambios que solicita que se extraigan. Por ejemplo, si Jessica quiere enviarle a John una solicitud de extracción y ha realizado dos confirmaciones en la rama del tema que acaba de enviar, puede ejecutar esto:

```
$ git request-pull origin/master myfork
The following changes since commit 1edee6b1d61823a2de3b09c160d7080b8d1b3a40:
Jessica Smith (1):
        Create new function

are available in the git repository at:

  git://githost/simplegit.git featureA

Jessica Smith (2):
      Add limit to log function
      Increase log output to 30 from 25

 lib/simplegit.rb |   10 +++++++++-
 1 files changed, 9 insertions(+), 1 deletions(-)
```

Esta salida se puede enviar al mantenedor: les dice de dónde se ramificó el trabajo, resume las confirmaciones e identifica de dónde se extraerá el nuevo trabajo.

En un proyecto para el que no eres el responsable de mantenimiento, generalmente es más fácil tener una rama como ```master``` siempre rastrear ```origin/master``` y hacer tu trabajo en ramas temáticas que puedes descartar fácilmente si son rechazadas. Tener temas de trabajo aislados en ramas temáticas también le facilita volver a basar su trabajo si la punta del repositorio principal se ha movido mientras tanto y sus confirmaciones ya no se aplican limpiamente. Por ejemplo, si desea enviar un segundo tema de trabajo al proyecto, no continúe trabajando en la rama del tema que acaba de subir; comience de nuevo desde la rama ```master``` del repositorio principal:

```
$ git checkout -b featureB origin/master
  ... work ...
$ git commit
$ git push myfork featureB
$ git request-pull origin/master myfork
  ... email generat```ed request pull to maintainer ...
$ git fetch origin
```

Ahora, cada uno de sus temas está contenido dentro de un silo, similar a una cola de parches, que puede reescribir, reajustar y modificar sin que los temas interfieran o dependan entre sí, así:

![](Images/img17.png)

Digamos que el responsable del proyecto ha incorporado un montón de otros parches y ha probado su primera rama, pero ya no se fusiona limpiamente. En este caso, puede intentar reajustar esa rama encima ```origin/master```, resolver los conflictos para el mantenedor y luego volver a enviar sus cambios:

```
$ git checkout featureA
$ git rebase origin/master
$ git push -f myfork featureA
```

Esto reescribe su historial para que ahora se vea como Confirmar historial después del trabajo featureA
  
![](Images/img18.png)

Debido a que cambió la base de la rama, debe especificar el comando  ```-f``` a su push para poder reemplazar la rama ```featureA``` en el servidor con una confirmación que no sea descendiente de ella. Una alternativa sería enviar este nuevo trabajo a una rama diferente en el servidor (quizás llamada ```featureAv2```).

Veamos un escenario más posible: el responsable de mantenimiento ha analizado el trabajo en su segunda rama y le gusta el concepto, pero le gustaría que cambiara un detalle de implementación. También aprovechará esta oportunidad para trasladar el trabajo para que se base en la rama master actual del proyecto . Comienza una nueva rama basada en la rama actual ```origin/master```, aplasta los cambios ```featureB``` allí, resuelve cualquier conflicto, realiza el cambio de implementación y luego empuja eso como una nueva rama:

```
$ git checkout -b featureBv2 origin/master
$ git merge --squash featureB
  ... change implementation ...
$ git commit
$ git push myfork featureBv2
```

La opción ```--squash``` toma todo el trabajo en la rama fusionada y lo aplasta en un conjunto de cambios produciendo el estado del repositorio como si hubiera ocurrido una fusión real, sin hacer realmente una confirmación de fusión. Esto significa que su futura confirmación solo tendrá un padre y le permitirá introducir todos los cambios de otra rama y luego realizar más cambios antes de grabar la nueva confirmación. Además, la opción ```--no-commit``` puede ser útil para retrasar la confirmación de fusión en caso del proceso de fusión predeterminado.

En este punto, puede notificar al mantenedor que ha realizado los cambios solicitados y que pueden encontrar esos cambios en su rama ```featureBv2```.

### Proyecto público por correo electrónico

Muchos proyectos han establecido procedimientos para aceptar parches; deberá verificar las reglas específicas para cada proyecto, ya que serán diferentes. Dado que hay varios proyectos más antiguos y más grandes que aceptan parches a través de una lista de correo de desarrolladores, veremos un ejemplo de eso ahora.

El flujo de trabajo es similar al caso de uso anterior: usted crea ramas temáticas para cada serie de parches en la que trabaja. La diferencia es cómo los envías al proyecto. En lugar de bifurcar (fork) el proyecto y pasar a su propia versión de escritura, genera versiones de correo electrónico de cada serie de confirmación y las envía por correo electrónico a la lista de correo del desarrollador:

```
$ git checkout -b topicA
  ... work ...
$ git commit
  ... work ...
$ git commit
```

Ahora tiene dos confirmaciones que desea enviar a la lista de correo. Usas ```git format-patch``` para generar los archivos con formato ```mbox``` que puedes enviar por correo electrónico a la lista; convierte cada confirmación en un mensaje de correo electrónico con la primera línea del mensaje de confirmación como asunto y el resto del mensaje más el parche que la confirmación introduce como el cuerpo. Lo bueno de esto es que la aplicación de un parche de un correo electrónico generado con ```format-patch``` conserva toda la información de confirmación correctamente.

```
$ git format-patch -M origin/master
0001-add-limit-to-log-function.patch
0002-increase-log-output-to-30-from-25.patch
```

El comando ```format-patch``` imprime los nombres de los archivos de parche que crea. El interruptor ```-M``` le dice a Git que busque cambios de nombre. Los archivos terminan luciendo así:

```
$ cat 0001-add-limit-to-log-function.patch
From 330090432754092d704da8e76ca5c05c198e71a8 Mon Sep 17 00:00:00 2001
From: Jessica Smith <jessica@example.com>
Date: Sun, 6 Apr 2008 10:17:23 -0700
Subject: [PATCH 1/2] Add limit to log function

Limit log functionality to the first 20

---
 lib/simplegit.rb |    2 +-
 1 files changed, 1 insertions(+), 1 deletions(-)

diff --git a/lib/simplegit.rb b/lib/simplegit.rb
index 76f47bc..f9815f1 100644
--- a/lib/simplegit.rb
+++ b/lib/simplegit.rb
@@ -14,7 +14,7 @@ class SimpleGit
   end

   def log(treeish = 'master')
-    command("git log #{treeish}")
+    command("git log -n 20 #{treeish}")
   end

   def ls_tree(treeish = 'master')
--
2.1.0
```

También puede editar estos archivos de parche para agregar más información para la lista de correo electrónico que no desea que aparezca en el mensaje de confirmación. Si agrega texto entre la línea --- y el comienzo del parche (la línea ```diff --git```), los desarrolladores pueden leerlo, pero el proceso de parcheo ignora ese contenido.

Para enviar esto por correo electrónico a una lista de correo, puede pegar el archivo en su programa de correo electrónico o enviarlo a través de un programa de línea de comandos. Pegar el texto a menudo causa problemas de formato, especialmente con clientes "más inteligentes" que no conservan las nuevas líneas y otros espacios en blanco de manera adecuada. Afortunadamente, Git proporciona una herramienta para ayudarlo a enviar parches formateados correctamente a través de IMAP, lo que puede ser más fácil para usted. Demostraremos cómo enviar un parche a través de Gmail, que resulta ser el agente de correo electrónico que mejor conocemos; puede leer instrucciones detalladas para varios programas de correo al final del archivo ```Documentation/SubmittingPatches``` mencionado anteriormente en el código fuente de Git.

Primero, debe configurar la sección ```imap``` en su archivo ```~/.gitconfig```. Puede establecer cada valor por separado con una serie de comandos ```git config```, o puede agregarlos manualmente, pero al final su archivo de configuración debería verse así:

```
[imap]
  folder = "[Gmail]/Drafts"
  host = imaps://imap.gmail.com
  user = user@gmail.com
  pass = YX]8g76G_2^sFbd
  port = 993
  sslverify = false
```

Si su servidor IMAP no usa SSL, las dos últimas líneas probablemente no sean necesarias y el valor de host será ```imap://``` en lugar de ```imaps://```. Una vez configurado, puede utilizar ```git imap-send``` para colocar la serie de parches en la carpeta Borradores del servidor IMAP especificado:

```
$ cat *.patch |git imap-send
Resolving imap.gmail.com... ok
Connecting to [74.125.142.109]:993... ok
Logging in...
sending 2 messages
100% (2/2) done
```

En este punto, debería poder ir a su carpeta Borradores, cambiar el campo Para a la lista de correo a la que está enviando el parche, posiblemente CC al mantenedor o persona responsable de esa sección, y envíelo.

También puede enviar los parches a través de un servidor SMTP. Como antes, puede establecer cada valor por separado con una serie de comandos ```git config```, o puede agregarlos manualmente en la sección de correo electrónico de su archivo:

```
[sendemail]
  smtpencryption = tls
  smtpserver = smtp.gmail.com
  smtpuser = user@gmail.com
  smtpserverport = 587
```  

Una vez hecho esto, puede usar ```git send-email``` para enviar sus parches:

```
$ git send-email *.patch
0001-add-limit-to-log-function.patch
0002-increase-log-output-to-30-from-25.patch

Emails will be sent from: Jessica Smith <jessica@example.com>
Who should the emails be sent to? jessica@example.com
Message-ID to be used as In-Reply-To for the first email? y
```

Luego, Git escupe un montón de información de registro similar a esta para cada parche que estás enviando:

```

  \line 'From: Jessica Smith <jessica@example.com>'
OK. Log says:
Sendmail: /usr/sbin/sendmail -i jessica@example.com
From: Jessica Smith <jessica@example.com>
To: jessica@example.com
Subject: [PATCH 1/2] Add limit to log function
Date: Sat, 30 May 2009 13:29:15 -0700
Message-Id: <1243715356-61726-1-git-send-email-jessica@example.com>
X-Mailer: git-send-email 1.6.2.rc1.20.g8c5b.dirty
In-Reply-To: <y>

```

**Consejo**

Para obtener ayuda sobre la configuración de su sistema y correo electrónico, más consejos y trucos, y una zona de pruebas para enviar un parche de prueba por correo electrónico, visite git-send-email.io.

**Resumen**

En esta sección, cubrimos múltiples flujos de trabajo y hablamos sobre las diferencias entre trabajar como parte de un equipo pequeño en proyectos de código cerrado y contribuir a un gran proyecto público. Sabe comprobar si hay errores de espacio en blanco antes de realizar la confirmación y puede escribir un gran mensaje de confirmación. Aprendió a formatear parches y enviarlos por correo electrónico a una lista de correo de desarrolladores. El tratamiento de las fusiones también se abordó en el contexto de los diferentes flujos de trabajo. Ahora está bien preparado para colaborar en cualquier proyecto.

A continuación, verá cómo trabajar la otra cara de la moneda: mantener un proyecto Git. Aprenderá a ser un dictador benevolente o un administrador de integración.

### Mantener un proyecto

Además de saber cómo contribuir eficazmente a un proyecto, es probable que necesite saber cómo mantener uno. Esto puede consistir en aceptar y aplicar parches generados ```format-patch``` y enviados por correo electrónico a usted, o integrar cambios en ramas remotas para repositorios que haya agregado como remotos a su proyecto. Ya sea que mantenga un repositorio canónico o quiera ayudar verificando o aprobando parches, necesita saber cómo aceptar el trabajo de la manera más clara para otros contribuyentes y sostenible para usted a largo plazo.

### Trabajar en ramas temáticas (topic branch)

Cuando esté pensando en integrar un nuevo trabajo, generalmente es una buena idea probarlo en una rama temática  , una rama temporal creada específicamente para probar ese nuevo trabajo. De esta manera, es fácil modificar un parche individualmente y dejarlo si no funciona hasta que tenga tiempo de volver a él. Si crea un nombre de rama simple basado en el tema del trabajo que va a probar, como ```ruby_client``` o algo similarmente descriptivo, puede recordarlo fácilmente si tiene que abandonarlo por un tiempo y volver más tarde. El mantenedor del proyecto Git también tiende a asignar un espacio de nombres a estas ramas, por ejemplo ```sc/ruby_client```, donde ```sc``` es la abreviatura de la persona que contribuyó con el trabajo. Como recordará, puede crear la rama basada en su rama ```master``` de esta manera:

```
$ git branch sc/ruby_client master
```

O, si también desea cambiarlo inmediatamente, puede usar la opción ```checkout -b```:

```
$ git checkout -b sc/ruby_client master
```

Ahora está listo para agregar el trabajo contribuido que recibió en esta rama temática y determinar si desea fusionarlo en sus ramas a más largo plazo.

### Aplicación de parches desde el correo electrónico

Si recibe un parche por correo electrónico que necesita integrar en su proyecto, debe aplicar el parche en su rama temática para evaluarlo. Hay dos formas de aplicar un parche enviado por correo electrónico: con ```git apply``` o con ```git am```.

### Aplicar un parche con apply

Si recibió el parche de alguien que lo generó con ```git diff``` o alguna variación del comando ```diff``` Unix (que no se recomienda; consulte la siguiente sección), puede aplicarlo con el comando ```git apply```. Suponiendo que guardó el parche en ```/tmp/patch-ruby-client.patch```, puede aplicar el parche de esta manera:

```
$ git apply /tmp/patch-ruby-client.patch
```

Esto modifica los archivos en su directorio de trabajo. Es casi idéntico a ejecutar un comando ```patch -p1``` para aplicar el parche, aunque es más paranoico y acepta menos coincidencias difusas que el parche. También maneja la adición, eliminación y cambio de nombre de archivos si se describen en el formato ```git diff```, lo que con patch no funciona. Por último, ```git apply``` hay un modelo de "aplicar todo o cancelar todo" en el que se aplica todo o no se aplica nada, mientras que ```patch``` pueden aplicar parches de forma parcial, dejando su directorio de trabajo en un estado extraño. ```git apply``` es en general mucho más conservador que ```patch```. No creará una confirmación para usted; después de ejecutarlo, debe realizar y confirmar los cambios introducidos manualmente.

También puede usar ```git apply``` para ver si un parche se aplica limpiamente antes de intentar aplicarlo realmente; puede ejecutarlo ```git apply --check``` con el parche:

```
$ git apply --check 0001-see-if-this-helps-the-gem.patch
error: patch failed: ticgit.gemspec:1
error: ticgit.gemspec: patch does not apply
```

Si no hay salida, entonces el parche debe aplicarse limpiamente. Este comando también sale con un estado distinto de cero si la verificación falla, por lo que puede usarlo en scripts si lo desea.

### Aplicar un parche con am

Si el colaborador es un usuario de Git y fue lo suficientemente bueno como para usar el comando ```format-patch``` para generar su parche, entonces su trabajo es más fácil porque el parche contiene información del autor y un mensaje de confirmación para usted. Si puede, anime a sus colaboradores a que utilicen ```format-patch``` en lugar  de ```diff``` para generar parches para usted. Solo debería tener que usar ```git appl``` y para parches heredados y cosas así.

Para aplicar un parche generado por ```format-patch```, usa ```git am``` (el comando se denomina ```am``` como se usa para "aplicar una serie de parches desde un buzón"). Técnicamente, ```git am``` está diseñado para leer un archivo mbox, que es un formato simple de texto sin formato para almacenar uno o más mensajes de correo electrónico en un archivo de texto. Se parece a esto:

```
From 330090432754092d704da8e76ca5c05c198e71a8 Mon Sep 17 00:00:00 2001
From: Jessica Smith <jessica@example.com>
Date: Sun, 6 Apr 2008 10:17:23 -0700
Subject: [PATCH 1/2] Add limit to log function

Limit log functionality to the first 20
```

Este es el comienzo de la salida del comando ```git format-patch``` que vio en la sección anterior; también representa un formato de correo electrónico mbox válido. Si alguien le ha enviado el parche por correo electrónico correctamente ```git send-email``` y lo descarga en un formato mbox, puede apuntar ```git am``` a ese archivo mbox y comenzará a aplicar todos los parches que vea. Si ejecuta un cliente de correo que puede guardar varios correos electrónicos en formato mbox, puede guardar toda la serie de parches en un archivo y luego usarlos ```git am``` para aplicarlos uno a la vez.

Sin embargo, si alguien cargó un archivo de parche generado a través ```git format-patch``` de un sistema de tickets o algo similar, puede guardar el archivo localmente y luego pasar ese archivo guardado en su disco ```git am``` para aplicarlo:

```
$ git am 0001-limit-log-function.patch
Applying: Add limit to log function
```

Puede ver que se aplicó limpiamente y creó automáticamente la nueva confirmación para usted. La información del autor se toma de los encabezados From y del correo electrónico Date, y el mensaje de la confirmación se toma del cuerpo Subject y (antes del parche) del correo electrónico. Por ejemplo, si este parche se aplicó desde el ejemplo de mbox anterior, la confirmación generada se vería así:

```
$ git log --pretty=fuller -1
commit 6c5e70b984a60b3cecd395edd5b48a7575bf58e0
Author:     Jessica Smith <jessica@example.com>
AuthorDate: Sun Apr 6 10:17:23 2008 -0700
Commit:     Scott Chacon <schacon@gmail.com>
CommitDate: Thu Apr 9 09:19:06 2009 -0700

   Add limit to log function

   Limit log functionality to the first 20
```

La información Commit indica la persona que aplicó el parche y la hora en que se aplicó. La información Author es la persona que creó originalmente el parche y cuándo se creó originalmente.

Pero es posible que el parche no se aplique limpiamente. Quizás su rama principal se ha alejado demasiado de la rama desde la que se construyó el parche, o el parche depende de otro parche que aún no ha aplicado. En ese caso, el proceso git am fallará y le preguntará qué desea hacer:

```
$ git am 0001-see-if-this-helps-the-gem.patch
Applying: See if this helps the gem
error: patch failed: ticgit.gemspec:1
error: ticgit.gemspec: patch does not apply
Patch failed at 0001.
When you have resolved this problem run "git am --resolved".
If you would prefer to skip this patch, instead run "git am --skip".
To restore the original branch and stop patching run "git am --abort".
```

Este comando coloca marcadores de conflicto en cualquier archivo con el que tenga problemas, al igual que una operación de fusión o rebase en conflicto. Resuelve este problema de la misma manera: edita el archivo para resolver el conflicto, prepara el nuevo archivo y luego ejecuta ```git am --resolved``` para continuar con el siguiente parche:

```
$ (fix the file)
$ git add ticgit.gemspec
$ git am --resolved
Applying: See if this helps the gem
```

Si desea que Git intente un poco más inteligentemente para resolver el conflicto, puede pasarle una opción ```-3```, lo que hace que Git intente una fusión de tres vías. Esta opción no está activada por defecto porque no funciona si la confirmación en la que el parche dice que se basó no está en su repositorio. Si tiene ese compromiso, si el parche se basó en un compromiso público, entonces la opción ```-3``` es generalmente mucho más inteligente sobre la aplicación de un parche conflictivo:

```
$ git am -3 0001-see-if-this-helps-the-gem.patch
Applying: See if this helps the gem
error: patch failed: ticgit.gemspec:1
error: ticgit.gemspec: patch does not apply
Using index info to reconstruct a base tree...
Falling back to patching base and 3-way merge...
No changes -- Patch already applied.
```

En este caso, sin la opción ```-3```, el parche se habría considerado un conflicto. Dado que se utilizó la opción -3, el parche se aplicó limpiamente.

Si está aplicando varios parches desde un mbox, también puede ejecutar el comando ```am``` en modo interactivo, que se detiene en cada parche que encuentra y le pregunta si desea aplicarlo:

```
$ git am -3 -i mbox
Commit Body is:
--------------------------
See if this helps the gem
--------------------------
Apply? [y]es/[n]o/[e]dit/[v]iew patch/[a]ccept all
```

Esto es bueno si tiene varios parches guardados, porque puede ver el parche primero si no recuerda cuál es, o no aplicar el parche si ya lo ha hecho.

Cuando todos los parches para su tema se aplican y se comprometen en su rama, puede elegir si desea integrarlos en una rama de ejecución más larga y cómo hacerlo.

### Comprobación de servidores remotos

Si su contribución provino de un usuario de Git que configuró su propio repositorio, introdujo una serie de cambios en él y luego le envió la URL al repositorio y el nombre de la rama remota en la que se encuentran los cambios, puede agregarlos como un remoto y fusiona localmente.

Por ejemplo, si Jessica le envía un correo electrónico diciendo que tiene una gran característica nueva en la  rama ```ruby-client``` de su repositorio, puede probarla agregando el servidor remoto y verificando esa rama localmente:

```
$ git remote add jessica git://github.com/jessica/myproject.git
$ git fetch jessica
$ git checkout -b rubyclient jessica/ruby-client
```

Si vuelve a enviarle un correo electrónico más tarde con otra rama que contenga otra gran función, podría hacerlo directamente fetch y checkout porque ya tiene la configuración remota.

Esto es más útil si trabaja con una persona de manera constante. Si alguien solo tiene un único parche para contribuir de vez en cuando, entonces aceptarlo por correo electrónico puede llevar menos tiempo que requerir que todos ejecuten su propio servidor y tener que agregar y eliminar servidores remotos continuamente para obtener algunos parches. También es poco probable que desee tener cientos de servidores remotos, cada uno para alguien que contribuye solo con un parche o dos. Sin embargo, los scripts y los servicios alojados pueden facilitar esto; depende en gran medida de cómo se desarrolle y cómo se desarrollen sus colaboradores.

La otra ventaja de este enfoque es que también obtiene el historial de las confirmaciones. Aunque puede tener problemas legítimos de fusión, sabe en qué parte de su historial se basa su trabajo; una combinación de tres vías adecuada es la predeterminada en lugar de tener que proporcionar un ```-3``` y esperar que el parche se haya generado a partir de una confirmación pública a la que tiene acceso.

Si no está trabajando con una persona de manera constante, pero aún desea extraer de ella de esta manera, puede proporcionar la URL del repositorio remoto al comando ```git pull```. Esto hace una extracción única y no guarda la URL como referencia remota:

```
$ git pull https://github.com/onetimeguy/project
From https://github.com/onetimeguy/project
 * branch            HEAD       -> FETCH_HEAD
Merge made by the 'recursive' strategy.
```

### Determinar qué se introduce

Ahora tiene una rama temática que contiene trabajos contribuidos. En este punto, puede determinar qué le gustaría hacer con él. Esta sección revisa un par de comandos para que pueda ver cómo puede usarlos para revisar exactamente lo que presentará si lo fusiona en su rama principal.

A menudo es útil obtener una revisión de todas las confirmaciones que están en esta rama pero que no están en su rama master. Puede excluir confirmaciones en la rama master agregando la opción ```--not``` antes del nombre de la rama. Esto hace lo mismo que el formato ```master..contrib``` que usamos anteriormente. Por ejemplo, si su colaborador le envía dos parches y usted crea una rama llamada contrib y aplica esos parches allí, puede ejecutar 
esto:

```
$ git log contrib --not master
commit 5b6235bd297351589efc4d73316f0a68d484f118
Author: Scott Chacon <schacon@gmail.com>
Date:   Fri Oct 24 09:53:59 2008 -0700

    See if this helps the gem

commit 7482e0d16d04bea79d0dba8988cc78df655f16a0
Author: Scott Chacon <schacon@gmail.com>
Date:   Mon Oct 22 19:38:36 2008 -0700

    Update gemspec to hopefully work better
```

Para ver qué cambios introduce cada confirmación, recuerde que puede pasar la opción  ```-p``` a ```git log``` y agregará la diferencia introducida a cada confirmación.

Para ver una diferencia completa de lo que sucedería si fusionara esta rama de tema con otra rama, es posible que deba usar un truco extraño para obtener los resultados correctos. Puede pensar en ejecutar esto:

```
$ git diff master
```

Este comando le da una diferencia, pero puede ser engañoso. Si su masterrama ha avanzado desde que creó la rama de tema a partir de ella, obtendrá resultados aparentemente extraños. Esto sucede porque Git compara directamente las instantáneas de la última confirmación de la rama del tema en la que se encuentra y la instantánea de la última confirmación en la masterrama. Por ejemplo, si ha agregado una línea en un archivo en la masterrama, una comparación directa de las instantáneas parecerá que la rama del tema va a eliminar esa línea.

Si master es un antepasado directo de su rama temática, esto no es un problema; pero si las dos historias han divergido, la diferencia parecerá que está agregando todas las cosas nuevas en su rama de tema y eliminando todo lo exclusivo de la masterrama.

Lo que realmente desea ver son los cambios agregados a la rama del tema, el trabajo que presentará si fusiona esta rama con master. Para hacerlo, haz que Git compare la última confirmación en tu rama de tema con el primer ancestro común que tiene con la masterrama.

Técnicamente, puede hacer eso descubriendo explícitamente el ancestro común y luego ejecutando su diff en él:

```
$ git merge-base contrib master
36c7dba2c95e6bbb78dfa822519ecfec6e1ca649
$ git diff 36c7db
```

o, más concisamente:

```
$ git diff $(git merge-base contrib master)
```

Sin embargo, ninguno de ellos es particularmente conveniente, por lo que Git proporciona otra abreviatura para hacer lo mismo: la sintaxis de tres puntos. En el contexto del comando ```git diff```, puede poner tres puntos después de otra rama para hacer una diffentre la última confirmación de la rama en la que se encuentra y su ancestro común con otra rama:

```
$ git diff master...contrib
```

Este comando le muestra solo el trabajo que ha introducido su rama de tema actual desde su antepasado común con master. Esa es una sintaxis muy útil para recordar.

### Integración del trabajo contribuido

Cuando todo el trabajo en su topic branch esté listo para integrarse en una rama más principal, la pregunta es cómo hacerlo. Además, ¿qué flujo de trabajo general desea utilizar para mantener su proyecto? Tiene varias opciones, por lo que cubriremos algunas de ellas.

### Fusionar flujos de trabajo

Un flujo de trabajo básico es simplemente fusionar todo ese trabajo directamente en su rama maste. En este escenario, tiene una rama master que contiene básicamente código estable. Cuando tiene trabajo en una rama de tema que cree que ha completado, o trabajo que otra persona ha contribuido y ha verificado, lo fusiona en su rama maestra, elimina esa rama de tema recién fusionada y repite.

Por ejemplo, si tenemos un repositorio con trabajo en dos ramas nombradas ```ruby_clienty php_client``` que se ve como Historial con varias ramas temáticas , y fusionamos ```ruby_client``` seguido de ```php_client```, su historial terminará luciendo como Después de la fusión de una rama temática.

![](Images/img20.png)

![](Images/img21.png)

Ese es probablemente el flujo de trabajo más simple, pero posiblemente puede ser problemático si se trata de proyectos más grandes o más estables en los que desea tener mucho cuidado con lo que introduce.

Si tiene un proyecto más importante, es posible que desee utilizar un ciclo de combinación de dos fases. En este escenario, tiene dos ramas de larga duración master y develop, en las que determina que master se actualiza solo cuando se corta una versión muy estable y todo el código nuevo se integra en la rama develop. Regularmente empuja ambas ramas al repositorio público. Cada vez que tenga una nueva rama de tema para fusionar ( antes de una fusión de rama de tema ), la fusionará en develop ( después de una fusión de rama de tema ); luego, cuando etiqueta una versión, avanza rápidamente master hasta donde develop esté la rama ahora estable ( después de una versión del proyecto ).

![](Images/img22.png)

![](Images/img23.png)

![](Images/img24.png4)

De esta manera, cuando las personas clonan el repositorio de su proyecto, pueden verificar la rama master para construir la última versión estable y mantenerse al día fácilmente, o pueden verificar develop, que es el contenido más vanguardista. También puede ampliar este concepto al tener una integraterama donde todo el trabajo se fusiona. Luego, cuando el código base de esa rama sea estable y pase las pruebas, lo fusionará en una developrama; y cuando ha demostrado ser estable por un tiempo, avanza rápidamente su masterrama.

### Flujos de trabajo de gran fusión

El proyecto Git tiene cuatro ramas de larga ejecución: master, next, y seen(anteriormente 'pu' - cambios propuestos) para un nuevo trabajo, y maintpara backports de mantenimiento. Cuando los colaboradores introducen un nuevo trabajo, se recopila en ramas temáticas en el repositorio del mantenedor de una manera similar a lo que hemos descrito (consulte Gestión de una serie compleja de ramas temáticas contribuidas en paralelo ). En este punto, los temas se evalúan para determinar si son seguros y están listos para el consumo o si necesitan más trabajo. Si son seguros, se fusionan nexty esa rama se empuja hacia arriba para que todos puedan probar los temas integrados juntos.

![](Images/img25.png)

Si los temas aún necesitan trabajo, se fusionan en su seenlugar. Cuando se determina que son totalmente estables, los temas se vuelven a fusionar en master. Las ramas nexty seenluego se reconstruyen a partir de master. Esto significa que mastercasi siempre avanza, nextse cambia de base ocasionalmente y seense cambia de base aún más a menudo:

![](Images/img26.png)

Cuando finalmente se fusiona una rama de tema master, se elimina del repositorio. El proyecto Git también tiene una maintrama que se bifurca desde la última versión para proporcionar parches retroactivos en caso de que se requiera una versión de mantenimiento. Por lo tanto, cuando clona el repositorio de Git, tiene cuatro ramas que puede consultar para evaluar el proyecto en diferentes etapas de desarrollo, dependiendo de qué tan avanzado quiera ser o cómo quiera contribuir; y el mantenedor tiene un flujo de trabajo estructurado para ayudarlos a examinar nuevas contribuciones. El flujo de trabajo del proyecto Git es especializado. Para entender esto claramente, puede consultar la guía de Git Maintainer .

### Flujos de trabajo de cambio de base y selección de cerezas

Otros mantenedores prefieren reajustar o seleccionar el trabajo contribuido en la parte superior de su masterrama, en lugar de fusionarlo, para mantener un historial principalmente lineal. Cuando ha trabajado en una rama de tema y ha determinado que desea integrarla, se mueve a esa rama y ejecuta el comando rebase para reconstruir los cambios en la parte superior de su rama actual master(o develop, etc.). Si eso funciona bien, puede adelantar su masterrama y terminará con un historial de proyecto lineal.

La otra forma de mover el trabajo introducido de una rama a otra es seleccionarlo. Una selección de cerebros en Git es como un cambio de base para una única confirmación. Toma el parche que se introdujo en una confirmación e intenta volver a aplicarlo en la rama en la que se encuentra actualmente. Esto es útil si tiene varias confirmaciones en una rama de tema y desea integrar solo una de ellas, o si solo tiene una confirmación en una rama de tema y prefiere seleccionarla en lugar de ejecutar rebase. Por ejemplo, suponga que tiene un proyecto que se parece a esto:

![](Images/img27.png)

Si desea ```e43a6``` incorporar el compromiso a su rama master, puede ejecutar:

```

$ git config --global rerere.enabled true
```

Ahora, cada vez que realice una fusión que resuelva conflictos, la resolución se registrará en la caché en caso de que la necesite en el futuro.

Si es necesario, puede interactuar con el caché rerere usando el comando ```git rerere```. Cuando se invoca solo, Git verifica su base de datos de resoluciones e intenta encontrar una coincidencia con los conflictos de fusión actuales y resolverlos (aunque esto se hace automáticamente si ```rerere.enabled```está configurado en true). También hay subcomandos para ver qué se grabará, para borrar una resolución específica de la caché y para borrar toda la caché. Cubriremos rerere con más detalle en Rerere .

### Etiquetar sus lanzamientos

Cuando haya decidido cortar una versión, probablemente desee asignar una etiqueta para poder volver a crear esa versión en cualquier momento en el futuro. Puede crear una nueva etiqueta como se explica en Conceptos básicos de Git . Si decide firmar la etiqueta como mantenedor, la etiqueta puede verse así:

```
$ git tag -s v1.5 -m 'my signed 1.5 tag'
You need a passphrase to unlock the secret key for
user: "Scott Chacon <schacon@gmail.com>"
1024-bit DSA key, ID F721C45A, created 2009-02-09
```

Si firma sus etiquetas, es posible que tenga el problema de distribuir la clave PGP pública utilizada para firmar sus etiquetas. El mantenedor del proyecto Git ha resuelto este problema al incluir su clave pública como un blob en el repositorio y luego agregar una etiqueta que apunta directamente a ese contenido. Para hacer esto, puede averiguar qué clave desea ejecutando ```gpg --list-keys```:

```
$ gpg --list-keys
/Users/schacon/.gnupg/pubring.gpg
---------------------------------
pub   1024D/F721C45A 2009-02-09 [expires: 2010-02-09]
uid                  Scott Chacon <schacon@gmail.com>
sub   2048g/45D02282 2009-02-09 [expires: 2010-02-09]
```

Luego, puede importar directamente la clave a la base de datos de Git exportándola y canalizándola ```git hash-object```, lo que escribe un nuevo blob con esos contenidos en Git y le devuelve el SHA-1 del blob:

```
$ gpg -a --export F721C45A | git hash-object -w --stdin
659ef797d181633c87ec71ac3f9ba29fe5775b92
```

Ahora que tiene el contenido de su clave en Git, puede crear una etiqueta que apunte directamente a ella especificando el nuevo valor SHA-1 que ```hash-object``` le dio el comando:

```
$ git tag -a maintainer-pgp-pub 659ef797d181633c87ec71ac3f9ba29fe5775b92
```

Si corre ```git push --tags```, la etiqueta ```maintainer-pgp-pub``` se compartirá con todos. Si alguien quiere verificar una etiqueta, puede importar directamente su clave PGP extrayendo el blob directamente de la base de datos e importándolo a GPG:

```
$ git show maintainer-pgp-pub | gpg --import
```

Pueden usar esa clave para verificar todas sus etiquetas firmadas. Además, si incluye instrucciones en el mensaje de la etiqueta, la ejecución ```git show <tag>``` le permitirá dar al usuario final instrucciones más específicas sobre la verificación de la etiqueta.

### Generación de un número de compilación
    
Debido a que Git no tiene números que aumentan monótonamente como 'v123' o el equivalente para cada confirmación, si desea tener un nombre legible por humanos para una confirmación, puede ejecutar git describe esa confirmación. En respuesta, Git genera una cadena que consta del nombre de la etiqueta más reciente anterior a esa confirmación, seguida del número de confirmaciones desde esa etiqueta, seguida finalmente de un valor SHA-1 parcial de la confirmación que se describe (con el prefijo de la letra "g" que significa Git):

```
$ git describe master
v1.6.2-rc1-20-g8c5b85c
```

De esta manera, puede exportar un snapshot o compilar y nombrarla de manera comprensible para las personas. De hecho, si compila Git a partir del código fuente clonado desde el repositorio de Git, obtendrá ```git --version``` algo parecido a esto. Si está describiendo una confirmación que ha etiquetado directamente, simplemente le da el nombre de la etiqueta.

De forma predeterminada, el comando ```git describe```  requiere etiquetas anotadas (etiquetas creadas con la bandera ```-a```  o ```-s``` ); Si también desea aprovechar las etiquetas ligeras (sin anotaciones), agregue la opción ```--tags``` al comando. También puede utilizar esta cadena como destino de un comando ```git checkout``` o ```git show```, aunque se basa en el valor SHA-1 abreviado al final, por lo que es posible que no sea válido para siempre. Por ejemplo, el kernel de Linux saltó recientemente de 8 a 10 caracteres para garantizar la exclusividad del objeto SHA-1, por lo que ```git describe``` se invalidaron los nombres de salida más antiguos .

### Preparando un lanzamiento

Ahora quieres lanzar una compilación. Una de las cosas que querrá hacer es crear un archivo de la última snapshot de su código para aquellas pobres almas que no usan Git. El comando para hacer esto es ```git archive```:

``` 
$ git archive master --prefix='project/' | gzip > `git describe master`.tar.gz
$ ls *.tar.gz
v1.6.2-rc1-20-g8c5b85c.tar.gz
``` 

Si alguien abre ese tarball, obtiene la última snapshot de su proyecto en un directorio ``` project``` . También puede crear un archivo zip de la misma manera, pero pasando la opción ``` --format=zip```  a ``` git archive``` :

```
$ git archive master --prefix='project/' --format=zip > `git describe master`.zip
```    

Ahora tiene un tarball agradable y un archivo zip del lanzamiento de su proyecto que puede cargar en su sitio web o enviar por correo electrónico a las personas.

### El Shortlog
    
Es hora de enviar un correo electrónico a su lista de correo de personas que desean saber qué está sucediendo en su proyecto. Una buena forma de obtener rápidamente una especie de registro de cambios de lo que se ha agregado a su proyecto desde su último lanzamiento o correo electrónico es usar el comando ```git shortlog```. Resume todas las confirmaciones en el rango que le da; por ejemplo, lo siguiente le brinda un resumen de todas las confirmaciones desde su última versión, si su última versión se llamó v1.0.1:

```
$ git shortlog --no-merges master --not v1.0.1
Chris Wanstrath (6):
      Add support for annotated tags to Grit::Tag
      Add packed-refs annotated tag support.
      Add Grit::Commit#to_patch
      Update version and History.txt
      Remove stray `puts`
      Make ls_tree ignore nils

Tom Preston-Werner (4):
      fix dates in history
      dynamic version method
      Version bump to 1.0.2
      Regenerated gemspec for version 1.0.2
```     

Obtiene un resumen limpio de todas las confirmaciones desde v1.0.1, agrupadas por autor, que puede enviar por correo electrónico a su lista.
