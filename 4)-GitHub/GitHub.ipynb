## GitHub

### Configuración y configuración de la cuenta

GitHub es el host más grande para repositorios de Git y es el punto central de colaboración para millones de desarrolladores y proyectos. Un gran porcentaje de todos los repositorios de Git están alojados en GitHub, y muchos proyectos de código abierto lo utilizan para el alojamiento de Git, el seguimiento de problemas, la revisión de código y otras cosas. Entonces, si bien no es una parte directa del proyecto de código abierto de Git, es muy probable que desee o necesite interactuar con GitHub en algún momento mientras usa Git profesionalmente.

Este capítulo trata sobre el uso eficaz de GitHub. Cubriremos el registro y la administración de una cuenta, la creación y el uso de repositorios de Git, los flujos de trabajo comunes para contribuir a los proyectos y aceptar contribuciones a los suyos, la interfaz programática de GitHub y muchos pequeños consejos para hacer su vida más fácil en general.

Si no está interesado en utilizar GitHub para alojar sus propios proyectos o colaborar con otros proyectos alojados en GitHub, puede pasar de forma segura a Git Tools.

**Nota**

Los planes de pago de GitHub incluyen herramientas y funciones avanzadas, así como mayores límites para los servicios gratuitos, pero no los cubriremos en este libro. Para obtener más información sobre los planes disponibles y su comparación, visite https://github.com/pricing .

### Sus direcciones de correo electrónico

La forma en que GitHub asigna su compromiso de Git con su usuario es mediante la dirección de correo electrónico. Si usa varias direcciones de correo electrónico en sus confirmaciones y desea que GitHub las vincule correctamente, debe agregar todas las direcciones de correo electrónico que ha usado a la sección Correos electrónicos de la sección de administración.

![](Images/img6.png)

En Agregar direcciones de correo electrónico podemos ver algunos de los diferentes estados que son posibles. ### Autenticación de dos factores

Finalmente, para mayor seguridad, definitivamente debe configurar la autenticación de dos factores o "2FA". La autenticación de dos factores es un mecanismo de autenticación que se está volviendo cada vez más popular recientemente para mitigar el riesgo de que su cuenta se vea comprometida si su contraseña es robada de alguna manera. Activarlo hará que GitHub te pida dos métodos diferentes de autenticación, de modo que si uno de ellos se ve comprometido, un atacante no podrá acceder a tu cuenta.

Puede encontrar la configuración de autenticación de dos factores en la pestaña Seguridad de la configuración de su cuenta.

![](Images/img7.png)

Si hace clic en el botón "Configurar autenticación de dos factores", lo llevará a una página de configuración donde puede elegir usar una aplicación de teléfono para generar su código secundario (una "contraseña de un solo uso basada en el tiempo"), o puede hacer que GitHub le envíe un código por SMS cada vez que necesite iniciar sesión.

Después de elegir qué método prefiere y seguir las instrucciones para configurar 2FA, su cuenta será un poco más segura y tendrá que proporcionar un código además de su contraseña cada vez que inicie sesión en GitHub.

### Contribuir a un proyecto

Ahora que nuestra cuenta está configurada, veamos algunos detalles que podrían ser útiles para ayudarlo a contribuir a un proyecto existente.

### Proyectos de bifurcación (fork)

Si desea contribuir a un proyecto existente al que no tiene acceso push, puede "bifurcar" (fork) el proyecto. Cuando “bifurca” (fork) un proyecto, GitHub hará una copia del proyecto que es completamente tuyo; vive en su espacio de nombres y puede acceder a él.

**Nota**

Históricamente, el término “bifurcación” ha tenido un contexto algo negativo, lo que significa que alguien tomó un proyecto de código abierto en una dirección diferente, a veces creando un proyecto competitivo y dividiendo a los contribuyentes. En GitHub, una "bifurcación" es simplemente el mismo proyecto en su propio espacio de nombres, lo que le permite realizar cambios en un proyecto públicamente como una forma de contribuir de una manera más abierta.

De esta manera, los proyectos no tienen que preocuparse por agregar usuarios como colaboradores para darles acceso push. Las personas pueden bifurcar un proyecto, impulsarlo y contribuir con sus cambios al repositorio original creando lo que se llama una **solicitud de extracción (pull request)**, que veremos a continuación. Esto abre un hilo de discusión con revisión de código, y el propietario y el colaborador pueden comunicarse sobre el cambio hasta que el propietario esté satisfecho con él, momento en el que el propietario puede fusionarlo.

Para bifurcar un proyecto, visite la página del proyecto y haga clic en el botón **"Bifurcar" (fork)** en la parte superior derecha de la página.

![](Images/img8.png)

Después de unos segundos, será llevado a la página de su nuevo proyecto, con su propia copia del código que se puede escribir.

### El flujo de GitHub

GitHub está diseñado en torno a un flujo de trabajo de colaboración particular, centrado en las **solicitudes de extracción (pull requests**). Este flujo funciona ya sea que esté colaborando con un equipo muy unido en un único repositorio compartido, o una empresa distribuida globalmente o una red de extraños que contribuyen a un proyecto a través de docenas de bifurcaciones. Se centra en el flujo de trabajo de las topic branch cubierto en Git Branching.

Así es como funciona generalmente:

1. Bifurca el proyecto.

2. Cree una topic branch a partir de master.

3. Comprométete (commit) a mejorar el proyecto.

4. Envíe esta rama (push y branch) a su proyecto de GitHub.

5. Abra una solicitud de extracción (pull request) en GitHub.

6. Discuta y, opcionalmente, continúe comprometiéndose (commit).

7. El propietario del proyecto fusiona (merge) o cierra la solicitud de extracción (pull request).

8. Sincroniza el  master actualizado (updated master) con tu bifurcación.

Este es básicamente el flujo de trabajo de Integration Manager cubierto en Integration-Manager Workflow, pero en lugar de usar el correo electrónico para comunicarse y revisar los cambios, los equipos usan las herramientas basadas en web de GitHub.

Veamos un ejemplo de cómo proponer un cambio en un proyecto de código abierto alojado en GitHub usando este flujo.

**Consejo**

Puede usar la herramienta CLI oficial de GitHub en lugar de la interfaz web de GitHub para la mayoría de las cosas. La herramienta se puede utilizar en sistemas Windows, MacOS y Linux. Vaya a la página de inicio de GitHub CLI para obtener instrucciones de instalación y el manual.

### Crear una solicitud de extracción (pull request)

Tony está buscando código para ejecutar en su microcontrolador programable Arduino y ha encontrado un gran archivo de programa en GitHub en https://github.com/schacon/blink.

![](Images/img9.png)

El único problema es que la frecuencia de parpadeo es demasiado rápida. Creemos que es mucho mejor esperar 3 segundos en lugar de 1 entre cada cambio de estado. Así que mejoremos el programa y envíelo de nuevo al proyecto como un cambio propuesto.

Primero, hacemos clic en el botón **'Bifurcación' (fork)** como se mencionó anteriormente para obtener nuestra propia copia del proyecto. Nuestro nombre de usuario aquí es "tonychacon", por lo que nuestra copia de este proyecto está en https://github.com/tonychacon/blinky ahí es donde podemos editarlo. Lo clonaremos localmente **(git clone)**, crearemos una **rama (topic branch) de tema**, cambiaremos el código y finalmente enviaremos ese cambio **(push)** a GitHub.

```
$ git clone https://github.com/tonychacon/blink (1)
Cloning into 'blink'...

$ cd blink
$ git checkout -b slow-blink (2)
Switched to a new branch 'slow-blink'

$ sed -i '' 's/1000/3000/' blink.ino (macOS) (3)
# If you're on a Linux system, do this instead:
# $ sed -i 's/1000/3000/' blink.ino (3)

$ git diff --word-diff (4)
diff --git a/blink.ino b/blink.ino
index 15b9911..a6cc5a5 100644
--- a/blink.ino
+++ b/blink.ino
@@ -18,7 +18,7 @@ void setup() {
// the loop routine runs over and over again forever:
void loop() {
  digitalWrite(led, HIGH);   // turn the LED on (HIGH is the voltage level)
  [-delay(1000);-]{+delay(3000);+}               // wait for a second
  digitalWrite(led, LOW);    // turn the LED off by making the voltage LOW
  [-delay(1000);-]{+delay(3000);+}               // wait for a second
}

$ git commit -a -m 'Change delay to 3 seconds' (5)
[slow-blink 5ca509d] Change delay to 3 seconds
 1 file changed, 2 insertions(+), 2 deletions(-)

$ git push origin slow-blink (6)
Username for 'https://github.com': tonychacon
Password for 'https://tonychacon@github.com':
Counting objects: 5, done.
Delta compression using up to 8 threads.
Compressing objects: 100% (3/3), done.
Writing objects: 100% (3/3), 340 bytes | 0 bytes/s, done.
Total 3 (delta 1), reused 0 (delta 0)
To https://github.com/tonychacon/blink
 * [new branch]      slow-blink -> slow-blink
``` 

1. Clone nuestra bifurcación (fork) del proyecto localmente.

2. Crea una rama temática (topic branch) descriptiva.

3. Realice nuestro cambio en el código.

4. Comprueba que el cambio sea bueno.

5. Confirme nuestro cambio en la rama del tema (topic branch).

6. Empuje nuestra nueva rama de tema de nuevo a nuestra bifurcación de GitHub.

Ahora, si volvemos a nuestra bifurcación en GitHub, podemos ver que GitHub notó que empujamos una nueva rama de tema hacia arriba y nos presenta un gran botón verde para verificar nuestros cambios y abrir una solicitud de extracción al proyecto original.

Alternativamente, puede ir a la página "Branches" en ```https://github.com/<user>/<project>/branches``` para ubicar su branch y abrir una nueva solicitud de extracción desde allí.
    
![](Images/img10.png)
    
Si hacemos clic en ese botón verde, veremos una pantalla que nos pide que le demos a nuestra **Pull Request** un título y una descripción. Casi siempre vale la pena esforzarse un poco en esto, ya que una buena descripción ayuda al propietario del proyecto original a determinar lo que estaba tratando de hacer, si los cambios propuestos son correctos y si aceptar los cambios mejoraría el proyecto original.

También vemos una lista de las confirmaciones en nuestra rama de tema (topic branch) que están "por delante" de la rama master (en este caso, solo una) y una diferencia unificada de todos los cambios que se realizarán si el propietario del proyecto fusiona esta rama.
    
![](Images/img11.png)
    
Cuando presiona el botón 'Crear solicitud de extracción' (pull request) en esta pantalla, el propietario del proyecto que bifurcó recibirá una notificación de que alguien está sugiriendo un cambio y se vinculará a una página que tiene toda esta información.

**Nota**

Aunque las solicitudes de extracción se usan comúnmente para proyectos públicos como este cuando el contribuyente tiene un cambio completo listo para realizar, también se usa a menudo en proyectos internos al comienzo del ciclo de desarrollo. Dado que puede seguir avanzando hacia la rama del tema incluso después de que se abre la solicitud de extracción, a menudo se abre temprano y se usa como una forma de iterar en el trabajo en equipo dentro de un contexto, en lugar de abrirse al final del proceso.

### Iterando en una solicitud de extracción

En este punto, el propietario del proyecto puede ver el cambio sugerido y fusionarlo (merge), rechazarlo o comentarlo. Digamos que le gusta la idea, pero preferiría un tiempo un poco más largo para que la luz esté apagada que encendida.

Cuando esta conversación puede tener lugar por correo electrónico en los flujos de trabajo presentados en Distributed Git , en GitHub esto sucede en línea. El propietario del proyecto puede revisar la diferencia unificada y dejar un comentario haciendo clic en cualquiera de las líneas.
    
![](Images/img12.png)
    
Una vez que el mantenedor hace este comentario, la persona que abrió la solicitud de extracción (y de hecho, cualquier otra persona que esté viendo el repositorio) recibirá una notificación. Repasaremos la personalización de esto más adelante, pero si tuviera las notificaciones por correo electrónico activadas, Tony recibiría un correo electrónico como este:

![](Images/img13.png)

Cualquiera también puede dejar comentarios generales sobre la solicitud de extracción. En la página de discusión de Pull Request podemos ver un ejemplo del propietario del proyecto comentando una línea de código y luego dejando un comentario general en la sección de discusión. Puede ver que los comentarios del código también se incorporan a la conversación.

![](Images/img14.png)
    
Ahora el colaborador puede ver lo que debe hacer para que se acepte el cambio. Afortunadamente, esto es muy sencillo. Donde por correo electrónico es posible que tenga que volver a publicar su serie y volver a enviarla a la lista de correo, con GitHub simplemente se compromete con la rama del tema (topic branch) nuevamente y empuja (push), lo que actualizará automáticamente la solicitud de extracción. En Pull Request final también puede ver que el comentario del código anterior se ha contraído en la Pull Request actualizada, ya que se realizó en una línea que desde entonces se ha modificado.

Agregar confirmaciones a una solicitud de extracción existente no activa una notificación, por lo que una vez que Tony ha presionado sus correcciones, decide dejar un comentario para informar al propietario del proyecto que realizó el cambio solicitado.

![](Images/img15.png)
    
Una cosa interesante a tener en cuenta es que si hace clic en la pestaña "Archivos modificados" en esta solicitud de extracción, obtendrá la diferencia "unificada", es decir, la diferencia agregada total que se introduciría en su rama principal si este topic branch en la que se fusionó. En términos ```git diff```, básicamente le muestra automáticamente ```git diff master…<branch>``` la rama en la que se basa esta solicitud de extracción. Consulte Determinar qué se introduce para obtener más información sobre este tipo de diferencia.

La otra cosa que notará es que GitHub verifica si la solicitud de extracción se fusiona limpiamente y proporciona un botón para hacer la combinación por usted en el servidor. Este botón solo aparece si tiene acceso de escritura al repositorio y es posible una combinación trivial. Si hace clic en él, GitHub realizará una fusión "sin avance rápido", lo que significa que incluso si la fusión pudiera ser un avance rápido, seguirá creando una confirmación de fusión.

Si lo prefiere, simplemente puede tirar de la rama hacia abajo y fusionarla localmente. Si fusiona esta rama en la rama master y la empuja a GitHub, la solicitud de extracción se cerrará automáticamente.

Este es el flujo de trabajo básico que utilizan la mayoría de los proyectos de GitHub. Se crean topic branchs, se abren solicitudes de extracción en ellas, se produce una discusión, posiblemente se realiza más trabajo en la rama y, finalmente, la solicitud se cierra o se fusiona.

**Nota**
No solo Forks

Es importante tener en cuenta que también puede abrir una solicitud de extracción entre dos ramas en el mismo repositorio. Si está trabajando en una función con alguien y ambos tienen acceso de escritura al proyecto, puede enviar una topic branch al repositorio y abrir una solicitud de extracción en la rama master de ese mismo proyecto para iniciar la revisión del código y discusión del proceso. No es necesario bifurcar.

### Solicitudes de extracción (pull request) avanzadas

Ahora que hemos cubierto los conceptos básicos de la contribución a un proyecto en GitHub, cubramos algunos consejos y trucos interesantes sobre las solicitudes de extracción para que pueda ser más eficaz en su uso.

### Solicitudes de extracción como parches

Es importante entender que muchos proyectos realmente no piensan en las solicitudes de extracción como colas de parches perfectos que deberían aplicarse limpiamente en orden, ya que la mayoría de los proyectos basados en listas de correo piensan en contribuciones de series de parches. La mayoría de los proyectos de GitHub piensan en las ramas de Pull Request como conversaciones iterativas en torno a un cambio propuesto, que culmina en una diferencia unificada que se aplica mediante la fusión.

Esta es una distinción importante, porque generalmente el cambio se sugiere antes de que se considere que el código es perfecto, lo cual es mucho más raro con las contribuciones de la serie de parches basados en listas de correo. Esto permite una conversación más temprana con los mantenedores para que llegar a la solución adecuada sea más un esfuerzo de la comunidad. Cuando se propone un código con una solicitud de extracción y los mantenedores o la comunidad sugieren un cambio, la serie de parches generalmente no se vuelve a publicar, sino que la diferencia se introduce como una nueva confirmación de la rama, lo que hace que la conversación avance con el contexto de la rama obra anterior intacta.

Por ejemplo, si regresa y mira de nuevo la Solicitud de extracción final , notará que el colaborador no volvió a basar su confirmación y envió otra Solicitud de extracción. En su lugar, agregaron nuevas confirmaciones y las enviaron a la rama existente. De esta manera, si vuelve atrás y mira este Pull Request en el futuro, puede encontrar fácilmente todo el contexto de por qué se tomaron las decisiones. Al presionar el botón "Combinar" en el sitio, se crea una confirmación de combinación que hace referencia a la solicitud de extracción para que sea fácil volver atrás e investigar la conversación original si es necesario.

### Mantenerse al día con Upstream

Si su Pull Request se vuelve obsoleta o no se fusiona limpiamente, querrá arreglarla para que el encargado de mantenimiento pueda fusionarla fácilmente. GitHub probará esto por ti y te informará en la parte inferior de cada Pull Request si la fusión es trivial o no.
    
![](Images/img16.png)
    
Si ve que algo como Pull Request no se fusiona limpiamente , querrá arreglar su rama para que se vuelva verde y el mantenedor no tenga que hacer trabajo adicional.

Tiene dos opciones principales para hacer esto. Puede cambiar la base de su rama sobre la rama de destino que sea (normalmente la rama master del repositorio que bifurcó), o puede fusionar la rama de destino en su rama.

La mayoría de los desarrolladores de GitHub elegirán hacer lo último, por las mismas razones que acabamos de comentar en la sección anterior. Lo que importa es el historial y la fusión final, por lo que el rebase no le brinda mucho más que un historial un poco más limpio y, a cambio, es mucho más difícil y propenso a errores.

Si desea fusionar en la rama de destino para que su Pull Request se pueda fusionar, debe agregar el repositorio original como un nuevo control remoto, buscarlo, fusionar la rama principal de ese repositorio en su topic branch, solucionar cualquier problema y finalmente enviarlo devuelva a la misma rama en la que abrió la solicitud de extracción.

Por ejemplo, digamos que en el ejemplo "tonychacon" que estábamos usando antes, el autor original hizo un cambio que crearía un conflicto en el Pull Request. Repasemos esos pasos.

```
$ git remote add upstream https://github.com/schacon/blink (1)

$ git fetch upstream (2)
remote: Counting objects: 3, done.
remote: Compressing objects: 100% (3/3), done.
Unpacking objects: 100% (3/3), done.
remote: Total 3 (delta 0), reused 0 (delta 0)
From https://github.com/schacon/blink
 * [new branch]      master     -> upstream/master

$ git merge upstream/master (3)
Auto-merging blink.ino
CONFLICT (content): Merge conflict in blink.ino
Automatic merge failed; fix conflicts and then commit the result.

$ vim blink.ino (4)
$ git add blink.ino
$ git commit
[slow-blink 3c8d735] Merge remote-tracking branch 'upstream/master' \
    into slower-blink

$ git push origin slow-blink (5)
Counting objects: 6, done.
Delta compression using up to 8 threads.
Compressing objects: 100% (6/6), done.
Writing objects: 100% (6/6), 682 bytes | 0 bytes/s, done.

To https://github.com/tonychacon/blink
   ef4725c..3c8d735  slower-blink -> slow-blink
```
   
1. Agregue el repositorio original como un nombre remoto upstream.

2. Busca el trabajo más reciente de ese control remoto.

3. Fusiona la rama principal de ese repositorio en tu rama temática.

4. Arregle el conflicto que ocurrió.

5. Vuelva a subir a la misma rama de tema.

Una vez que lo haga, la solicitud de extracción se actualizará automáticamente y se volverá a verificar para ver si se fusiona limpiamente.

![](Images/img17.png)

Una de las mejores cosas de Git es que puedes hacerlo de forma continua. Si tiene un proyecto de larga duración, puede fusionar fácilmente desde la rama de destino una y otra vez y solo tendrá que lidiar con los conflictos que han surgido desde la última vez que se fusionó, lo que hace que el proceso sea muy manejable.

Si absolutamente desea volver a colocar la base en la rama para limpiarla, ciertamente puede hacerlo, pero se recomienda encarecidamente no forzar el empuje sobre la rama en la que ya está abierta la Solicitud de extracción. Si otras personas lo han derribado y han trabajado más en él, se encontrará con todos los problemas descritos en Los peligros del cambio de base . En su lugar, empuje la rama rebasada a una nueva rama en GitHub y abra una nueva solicitud de extracción que haga referencia a la anterior, luego cierre la original.

### Referencias

Su siguiente pregunta puede ser "¿Cómo puedo hacer referencia a la solicitud de extracción anterior?". Resulta que hay muchas, muchas formas de hacer referencia a otras cosas en casi cualquier lugar donde puedas escribir en GitHub.

Comencemos con cómo hacer una referencia cruzada a otra solicitud de extracción o un problema. A todas las solicitudes de extracción y los problemas se les asignan números y son únicos dentro del proyecto. Por ejemplo, no puede tener Pull Request # 3 y Issue # 3. Si desea hacer referencia a cualquier Pull Request o Issue de cualquier otro, simplemente puede poner ```#<num>``` cualquier comentario o descripción. También puede ser más específico si el problema o la solicitud de extracción se encuentra en otro lugar; escriba ```username#<num>``` si se refiere a un problema o una solicitud de extracción en una bifurcación del repositorio en el que se encuentra, o ```username/repo#<num>``` para hacer referencia a algo en otro repositorio.

Veamos un ejemplo. Digamos que cambiamos la base de la rama en el ejemplo anterior, creamos una nueva solicitud de extracción para ella y ahora queremos hacer referencia a la solicitud de extracción anterior de la nueva. También queremos hacer referencia a un problema en la bifurcación del repositorio y un problema en un proyecto completamente diferente. Podemos completar la descripción como referencias cruzadas en una solicitud de extracción .

![](Images/img18.png)

Cuando enviemos esta solicitud de extracción, veremos todo eso representado como referencias cruzadas representadas en una solicitud de extracción .

![](Images/img19.png)

Tenga en cuenta que la URL completa de GitHub que colocamos allí se acortó a solo la información necesaria.

Ahora, si Tony retrocede y cierra la solicitud de extracción original, podemos ver que al mencionarla en la nueva, GitHub ha creado automáticamente un evento de retroceso en la línea de tiempo de la solicitud de extracción. Esto significa que cualquier persona que visite este Pull Request y vea que está cerrada puede vincular fácilmente a la que la reemplazó. El enlace se verá algo así como Enlace a la nueva solicitud de extracción en la línea de tiempo cerrada de la solicitud de extracción .

![](Images/img20.png)

Además de los números de emisión, también puede hacer referencia a una confirmación específica por SHA-1. Debe especificar un SHA-1 completo de 40 caracteres, pero si GitHub lo ve en un comentario, se vinculará directamente a la confirmación. Nuevamente, puede hacer referencia a confirmaciones en bifurcaciones u otros repositorios de la misma manera que lo hizo con los problemas.

### Markdown con sabor a GitHub 

Vincular a otros problemas es solo el comienzo de cosas interesantes que puede hacer con casi cualquier cuadro de texto en GitHub. En las descripciones de problemas y solicitudes de extracción, comentarios, comentarios de código y más, puede usar lo que se llama "GitHub Flavored Markdown". Markdown es como escribir en texto sin formato pero que se representa de forma rica.

Consulte Un ejemplo de GitHub Flavored Markdown tal como está escrito y representado para ver un ejemplo de cómo se pueden escribir comentarios o texto y luego representarlos usando Markdown.

![](Images/img21.png)

La versión GitHub de Markdown agrega más cosas que puede hacer más allá de la sintaxis básica de Markdown. Todos estos pueden ser realmente útiles al crear comentarios o descripciones útiles de Pull Request o Issue.

### Listas de tareas

La primera función de Markdown específica de GitHub realmente útil, especialmente para usar en solicitudes de extracción, es la lista de tareas. Una lista de tareas es una lista de casillas de verificación de las cosas que desea hacer. Ponerlos en un Issue o Pull Request normalmente indica cosas que desea hacer antes de considerar que el elemento está completo.

Puede crear una lista de tareas como esta:

- [X] Write the code
- [ ] Write all the tests
- [ ] Document the code

Si incluimos esto en la descripción de nuestra solicitud de extracción o problema, lo veremos representado como listas de tareas representadas en un comentario de Markdown.

![](Images/img22.png)

Esto se usa a menudo en las solicitudes de extracción para indicar todo lo que le gustaría hacer en la rama antes de que la solicitud de extracción esté lista para fusionarse. La parte realmente interesante es que simplemente puede hacer clic en las casillas de verificación para actualizar el comentario; no tiene que editar el Markdown directamente para marcar las tareas.

Además, GitHub buscará listas de tareas en sus problemas y solicitudes de extracción y las mostrará como metadatos en las páginas que las enumeran. Por ejemplo, si tiene una solicitud de extracción con tareas y observa la página de descripción general de todas las solicitudes de extracción, puede ver qué tan lejos está. Esto ayuda a las personas a dividir las solicitudes de extracción en subtareas y ayuda a otras personas a realizar un seguimiento del progreso de la rama. Puede ver un ejemplo de esto en el resumen de la lista de tareas en la lista de solicitudes de extracción.

![](Images/img23.png)

Estos son increíblemente útiles cuando abre una solicitud de extracción con anticipación y la usa para realizar un seguimiento de su progreso a través de la implementación de la función.

### Fragmentos de código

También puede agregar fragmentos de código a los comentarios. Esto es especialmente útil si desea presentar algo que podría intentar hacer antes de implementarlo como una confirmación en su rama. Esto también se usa a menudo para agregar código de ejemplo de lo que no está funcionando o lo que este Pull Request podría implementar.

Para agregar un fragmento de código, debe "ponerlo" con comillas invertidas.

```java
for(int i=0 ; i < 5 ; i++)
{
   System.out.println("i is : " + i);
}
```

### Mantenga actualizado su repositorio público de GitHub

Una vez que haya bifurcado un repositorio de GitHub, su repositorio (su "bifurcación") existe independientemente del original. En particular, cuando el repositorio original tiene nuevas confirmaciones, GitHub te informa con un mensaje como:

```
This branch is 5 commits behind progit:master.
```

Pero su repositorio de GitHub nunca será actualizado automáticamente por GitHub; esto es algo que debe hacer usted mismo. Afortunadamente, esto es muy fácil de hacer.

Una posibilidad para hacer esto no requiere configuración. Por ejemplo, si se bifurcó https://github.com/progit/progit2.git, puede mantener su branch master actualizada de esta manera:

```
$ git checkout master (1)
$ git pull https://github.com/progit/progit2.git (2)
$ git push origin master (3)
```

1. Si estaba en otra branch, regrese a master.

2. Obtener cambios de https://github.com/progit/progit2.gity fusionarlos en master.

3. Empuje su rama master a origin.

Esto funciona, pero es un poco tedioso tener que deletrear la URL de búsqueda cada vez. Puede automatizar este trabajo con un poco de configuración:

```
$ git remote add progit https://github.com/progit/progit2.git (1)
$ git fetch progit (2)
$ git branch --set-upstream-to=progit/master master (3)
$ git config --local remote.pushDefault origin (4)
```

1. Agregue el repositorio de origen y asígnele un nombre. Aquí, he elegido llamarlo progit.

2. Obtenga una referencia sobre las ramas de progit, en particular master.

3. Configure su masterrama para buscar desde el progitcontrol remoto.

4. Defina el repositorio de inserción predeterminado en origin.

Una vez hecho esto, el flujo de trabajo se vuelve mucho más simple:

```
$ git checkout master (1)
$ git pull (2)
$ git push (3)
```

1. Si estaba en otra branch, regrese a master.

2. Obtener cambios de progit y fusionar cambios en master.

3. Empuje su rama master a origin.

Este enfoque puede ser útil, pero no está exento de inconvenientes. Git felizmente hará este trabajo por usted en silencio, pero no le advertirá si se compromete a master, tira de progit y luego empuja a origin - todas esas operaciones son válidas con esta configuración. Por lo tanto, tendrá que tener cuidado de no comprometerse directamente con master, ya que esa rama pertenece efectivamente al repositorio ascendente.

### Administrar solicitudes de extracción (pull request)

Ahora que tiene un proyecto con algo de código y tal vez incluso algunos colaboradores que también tienen acceso push, repasemos qué hacer cuando obtenga una solicitud de extracción (pull request) usted mismo.

Las solicitudes de extracción pueden provenir de una rama en una bifurcación (fork )de su repositorio o pueden provenir de otra rama en el mismo repositorio. La única diferencia es que los que están en una bifurcación (fork) a menudo provienen de personas a las que no puedes hacer push a su branch y ellos no pueden hacer push a la tuya, mientras que con las solicitudes de extracción (pull request) internas generalmente ambas partes pueden acceder a la branch.

Para estos ejemplos, supongamos que es "tonychacon" y ha creado un nuevo proyecto de código Arduino llamado "fade".

### Notificaciónes de Correo Electrónico

Alguien viene, hace un cambio en su código y le envía una solicitud de extracción. Debería recibir un correo electrónico que le notifique sobre la nueva solicitud de extracción y debería verse como una notificación por correo electrónico de una nueva solicitud de extracción .
    
![](Images/img34.png)

Hay algunas cosas a tener en cuenta sobre este correo electrónico. Le dará un pequeño diffstat: una lista de archivos que han cambiado en la solicitud de extracción y por cuánto. Le brinda un enlace a la solicitud de extracción en GitHub. También le brinda algunas URL que puede usar desde la línea de comandos.

Si nota la línea que dice ```git pull <url> patch-1```, esta es una forma sencilla de fusionar en una rama remota sin tener que agregar una remota. Repasamos esto rápidamente en Verificación de sucursales remotas. Si lo desea, puede crear y cambiar a una rama de tema y luego ejecutar este comando para fusionar los cambios de Pull Request.

Las otras URL interesantes son las URL .diff y .patch, que, como puede adivinar, proporcionan versiones unificadas de diferencias y parches de la solicitud de extracción. Técnicamente, podría fusionar el trabajo de Pull Request con algo como esto:

```
$ curl https://github.com/tonychacon/fade/pull/1.patch | git am
```

### Referencias de solicitud de extracción

Si está lidiando con muchas solicitudes de extracción y no desea agregar un montón de servidores remotos o hacer extracciones únicas cada vez, hay un buen truco que GitHub le permite hacer. Este es un truco un poco avanzado y repasaremos los detalles de esto un poco más en The Refspec, pero puede ser bastante útil.

GitHub en realidad anuncia las ramas de Pull Request para un repositorio como una especie de pseudo-ramas en el servidor. De forma predeterminada, no los obtiene cuando clona, pero están allí de una manera oculta y puede acceder a ellos con bastante facilidad.

Para demostrar esto, usaremos un comando de bajo nivel (a menudo denominado comando de "plomería", sobre el cual leeremos más en Plomería y porcelana) llamado ```ls-remote```. Este comando generalmente no se usa en las operaciones diarias de Git, pero es útil para mostrarnos qué referencias están presentes en el servidor.

Si ejecutamos este comando contra el repositorio "blink" que estábamos usando anteriormente, obtendremos una lista de todas las ramas y etiquetas y otras referencias en el repositorio.

```
$ git ls-remote https://github.com/schacon/blink
10d539600d86723087810ec636870a504f4fee4d	HEAD
10d539600d86723087810ec636870a504f4fee4d	refs/heads/master
6a83107c62950be9453aac297bb0193fd743cd6e	refs/pull/1/head
afe83c2d1a70674c9505cc1d8b7d380d5e076ed3	refs/pull/1/merge
3c8d735ee16296c242be7a9742ebfbc2665adec1	refs/pull/2/head
15c9f4f80973a2758462ab2066b6ad9fe8dcf03d	refs/pull/2/merge
a5a7751a33b7e86c5e9bb07b26001bb17d775d1a	refs/pull/4/head
31a45fc257e8433c8d8804e3e848cf61c9d3166c	refs/pull/4/merge
```

Por supuesto, si está en su repositorio y ejecuta ```git ls-remote origin``` o cualquier servidor remoto que desee verificar, le mostrará algo similar a esto.

Si el repositorio está en GitHub y tiene solicitudes de extracción que se han abierto, obtendrá estas referencias con el prefijo ```refs/pull/```. Estas son básicamente ramas, pero como no están debajo ```refs/heads/```, no las obtiene normalmente cuando clona o recupera del servidor; el proceso de búsqueda las ignora normalmente.

Hay dos referencias por solicitud de extracción: la que termina en ```/head``` puntos exactamente a la misma confirmación que la última confirmación en la rama Solicitud de extracción. Así que si alguien abre una solicitud de extracción en nuestro repositorio y su rama se nombra ```bug-fix``` y apunta a comprometerse a5a775, a continuación, en nuestro repositorio no vamos a tener una  rama bug-fix (ya que está en su fork), pero nos vamos a tener ```pull/<pr#>/head``` que apunta a a5a775. Esto significa que podemos desplegar fácilmente todas las ramas de Pull Request de una sola vez sin tener que agregar un montón de servidores remotos.

Ahora, puede hacer algo como buscar la referencia directamente.

```
$ git fetch origin refs/pull/958/head
From https://github.com/libgit2/libgit2
 * branch            refs/pull/958/head -> FETCH_HEAD
```

Esto le dice a Git, "Conéctese al servidor remoto origin y descargue la referencia nombrada ```refs/pull/958/head```". Git obedece felizmente y descarga todo lo que necesita para construir esa referencia, y pone un puntero al compromiso que desea debajo ```.git/FETCH_HEAD```. Puede seguir ese ```git merge FETCH_HEAD``` en una rama en la que desee probarlo, pero ese mensaje de confirmación de fusión parece un poco extraño. Además, si está revisando muchas solicitudes de extracción, esto se vuelve tedioso.

También hay una forma de obtener todas las solicitudes de extracción y mantenerlas actualizadas siempre que se conecte al servidor remoto. Abra ```.git/config``` en su editor favorito y busque el servidor remoto origin. Debería verse un poco así:

```
[remote "origin"]
    url = https://github.com/libgit2/libgit2
    fetch = +refs/heads/*:refs/remotes/origin/*
```    

Esa línea que comienza con ```fetch =``` es una "refspec". Es una forma de mapear nombres en el servidor remoto con nombres en su directorio local ```.git```. Este en particular le dice a Git, "las cosas en el servidor remoto que están debajo ```refs/heads``` deben ir en mi repositorio local debajo ```refs/remotes/origin```". Puede modificar esta sección para agregar otra especificación de referencia:

```
[remote "origin"]
    url = https://github.com/libgit2/libgit2.git
    fetch = +refs/heads/*:refs/remotes/origin/*
    fetch = +refs/pull/*/head:refs/remotes/origin/pr/*
```    

Esa última línea le dice a Git, "Todas las referencias que se parecen ```refs/pull/123/head``` deben almacenarse localmente como ```refs/remotes/origin/pr/123```". Ahora, si guarda ese archivo y hace ```git fetch```:

```
$ git fetch
# …
 * [new ref]         refs/pull/1/head -> origin/pr/1
 * [new ref]         refs/pull/2/head -> origin/pr/2
 * [new ref]         refs/pull/4/head -> origin/pr/4
# …
```

Ahora todas las solicitudes de extracción remotas se representan localmente con referencias que actúan de forma muy similar a las ramas de seguimiento; son de solo lectura y se actualizan cuando realiza una búsqueda. Esto hace que sea muy fácil probar el código de una solicitud de extracción localmente:

```
$ git checkout pr/2
Checking out files: 100% (3769/3769), done.
Branch pr/2 set up to track remote branch pr/2 from origin.
Switched to a new branch 'pr/2'
```

Los ojos de águila entre ustedes notarían el head al final de la parte remota de la especificación de referencia. También hay una referencia ```refs/pull/#/merge``` en el lado de GitHub, que representa la confirmación que resultaría si presiona el botón "fusionar" en el sitio. Esto puede permitirle probar la combinación incluso antes de presionar el botón.

### Hooks

Si necesita algo más específico o desea integrarse con un servicio o sitio que no está incluido en esta lista, puede usar el sistema de Hooks más genérico. Los enlaces del repositorio de GitHub son bastante simples. Usted especifica una URL y GitHub publicará una carga útil HTTP en esa URL en cualquier evento que desee.

Generalmente, la forma en que esto funciona es que puede configurar un pequeño servicio web para escuchar una carga útil de Hook de GitHub y luego hacer algo con los datos cuando se reciben.

Para habilitar un enlace, haga clic en el botón "Agregar webhook" en la sección de configuración de Servicios y enlaces . Esto lo llevará a una página que se parece a la configuración de un enlace web.

![](Images/img51.png)

La configuración de un web hook es bastante simple. En la mayoría de los casos, simplemente ingresa una URL y una clave secreta y presiona "Agregar webhook". Hay algunas opciones para los eventos para los que desea que GitHub le envíe una carga útil; el valor predeterminado es solo obtener una carga útil para el evento  push, cuando alguien inserta un nuevo código en cualquier rama de su repositorio.

Veamos un pequeño ejemplo de un servicio web que puede configurar para manejar un enlace web. Usaremos el marco web Ruby Sinatra ya que es bastante conciso y debería poder ver fácilmente lo que estamos haciendo.

Digamos que queremos recibir un correo electrónico si una persona específica empuja (push) a una rama específica de nuestro proyecto modificando un archivo específico. Podríamos hacer eso con bastante facilidad con un código como este:

```
require 'sinatra'
require 'json'
require 'mail'

post '/payload' do
  push = JSON.parse(request.body.read) # parse the JSON

  # gather the data we're looking for
  pusher = push["pusher"]["name"]
  branch = push["ref"]

  # get a list of all the files touched
  files = push["commits"].map do |commit|
    commit['added'] + commit['modified'] + commit['removed']
  end
  files = files.flatten.uniq

  # check for our criteria
  if pusher == 'schacon' &&
     branch == 'ref/heads/special-branch' &&
     files.include?('special-file.txt')

    Mail.deliver do
      from     'tchacon@example.com'
      to       'tchacon@example.com'
      subject  'Scott Changed the File'
      body     "ALARM"
    end
  end
end
```

Aquí tomamos la carga útil JSON que nos entrega GitHub y buscamos quién la empujó (push), a qué rama empujaron y qué archivos se tocaron en todas las confirmaciones que se enviaron. Luego lo comparamos con nuestros criterios y enviamos un correo electrónico si coincide.

Para desarrollar y probar algo como esto, tiene una buena consola de desarrollador en la misma pantalla donde configuró el enlace. Puede ver las últimas entregas que GitHub ha intentado realizar para ese webhook. Para cada Hooks, puede profundizar en cuándo se entregó, si fue exitoso y el cuerpo y los encabezados tanto de la solicitud como de la respuesta. Esto hace que sea increíblemente fácil probar y depurar sus Hooks.

![](Images/img52.png)

La otra gran característica de esto es que puede volver a entregar cualquiera de las cargas útiles para probar su servicio fácilmente.

Para obtener más información sobre cómo escribir webhooks y todos los diferentes tipos de eventos que puede escuchar, vaya a la documentación para desarrolladores de GitHub en https://developer.github.com/webhooks/.

### La API de GitHub

Los servicios y los hooks le brindan una forma de recibir notificaciones automáticas sobre eventos que suceden en sus repositorios, pero ¿qué sucede si necesita más información sobre estos eventos? ¿Qué sucede si necesita automatizar algo como agregar colaboradores o etiquetar problemas?

Aquí es donde la API de GitHub resulta útil. GitHub tiene toneladas de puntos finales de API para hacer casi cualquier cosa que pueda hacer en el sitio web de forma automatizada. En esta sección, aprenderemos cómo autenticarnos y conectarnos a la API, cómo comentar un problema y cómo cambiar el estado de un **Pull Request** a través de la API.

### Uso básico

Lo más básico que puede hacer es una simple solicitud GET en un punto final que no requiere autenticación. Esto podría ser información de usuario o de solo lectura en un proyecto de código abierto. Por ejemplo, si queremos saber más sobre un usuario llamado "schacon", podemos ejecutar algo como esto:

```
$ curl https://api.github.com/users/schacon
{
  "login": "schacon",
  "id": 70,
  "avatar_url": "https://avatars.githubusercontent.com/u/70",
# …
  "name": "Scott Chacon",
  "company": "GitHub",
  "following": 19,
  "created_at": "2008-01-27T17:19:28Z",
  "updated_at": "2014-06-10T02:37:23Z"
}
```

Hay toneladas de puntos finales como este para obtener información sobre organizaciones, proyectos, problemas, confirmaciones, casi cualquier cosa que pueda ver públicamente en GitHub. Incluso puede usar la API para representar Markdown arbitrario o encontrar una plantilla .gitignore.

```
$ curl https://api.github.com/gitignore/templates/Java
{
  "name": "Java",
  "source": "*.class

# Mobile Tools for Java (J2ME)
.mtj.tmp/

# Package Files #
*.jar
*.war
*.ear

# virtual machine crash logs, see https://www.java.com/en/download/help/error_hotspot.xml
hs_err_pid*
"
}
```

### Comentar un problema (Issue)

Sin embargo, si desea realizar una acción en el sitio web, como comentar un problema (Issue) o una solicitud de extracción (pull request), o si desea ver o interactuar con contenido privado, deberá autenticarse.

Hay varias formas de autenticarse. Puede usar la autenticación básica solo con su nombre de usuario y contraseña, pero generalmente es una mejor idea usar un token de acceso personal. Puede generar esto desde la pestaña "Aplicaciones" de su página de configuración.

![](Images/img53.png)

```
$ curl -H "Content-Type: application/json" \
       -H "Authorization: token TOKEN" \
       --data '{"body":"A new comment, :+1:"}' \
       https://api.github.com/repos/schacon/blink/issues/6/comments
{
  "id": 58322100,
  "html_url": "https://github.com/schacon/blink/issues/6#issuecomment-58322100",
  ...
  "user": {
    "login": "tonychacon",
    "id": 7874698,
    "avatar_url": "https://avatars.githubusercontent.com/u/7874698?v=2",
    "type": "User",
  },
  "created_at": "2014-10-08T07:48:19Z",
  "updated_at": "2014-10-08T07:48:19Z",
  "body": "A new comment, :+1:"
}
```

Ahora, si va a ese problema, puede ver el comentario que acabamos de publicar con éxito como en un comentario publicado desde la API de GitHub .

![](Images/img54.png)

Puede usar la API para hacer casi cualquier cosa que pueda hacer en el sitio web: crear y establecer hitos, asignar personas a problemas y solicitudes de extracción, crear y cambiar etiquetas, acceder a datos de confirmación, crear nuevas confirmaciones y branchs, abrir, cerrar o fusionar Pull Requests, creación y edición de equipos, comentando líneas de código en un Pull Request, buscando en el sitio y así sucesivamente.

### Cambiar el estado de una solicitud de extracción (Pull request)

Hay un ejemplo final que veremos, ya que es realmente útil si está trabajando con solicitudes de extracción. Cada confirmación puede tener uno o más estados asociados y hay una API para agregar y consultar ese estado.

La mayoría de los servicios de prueba e integración continua hacen uso de esta API para reaccionar a los empujes (push) probando el código que se envió y luego informan si esa confirmación ha pasado todas las pruebas. También puede usar esto para verificar si el mensaje de confirmación tiene el formato correcto, si el remitente siguió todas sus pautas de contribución, si la confirmación se firmó de manera válida, cualquier cantidad de cosas.

Supongamos que configura un webhook en su repositorio que llega a un pequeño servicio web que busca una cadena  Signed-off-by (Firmado por) en el mensaje de confirmación.

```
require 'httparty'
require 'sinatra'
require 'json'

post '/payload' do
  push = JSON.parse(request.body.read) # parse the JSON
  repo_name = push['repository']['full_name']

  # look through each commit message
  push["commits"].each do |commit|

    # look for a Signed-off-by string
    if /Signed-off-by/.match commit['message']
      state = 'success'
      description = 'Successfully signed off!'
    else
      state = 'failure'
      description = 'No signoff found.'
    end

    # post status to GitHub
    sha = commit["id"]
    status_url = "https://api.github.com/repos/#{repo_name}/statuses/#{sha}"

    status = {
      "state"       => state,
      "description" => description,
      "target_url"  => "http://example.com/how-to-signoff",
      "context"     => "validate/signoff"
    }
    HTTParty.post(status_url,
      :body => status.to_json,
      :headers => {
        'Content-Type'  => 'application/json',
        'User-Agent'    => 'tonychacon/signoff',
        'Authorization' => "token #{ENV['TOKEN']}" }
    )
  end
end
```

Con suerte, esto es bastante sencillo de seguir. En este controlador de enlace web, examinamos cada confirmación que se acaba de enviar, buscamos la cadena 'Firmado por' en el mensaje de confirmación y, finalmente, POST a través de HTTP al ```/repos/<user>/<repo>/statuses/<commit_sha>``` punto final de la API con el estado.

En este caso, puede enviar un estado ('éxito', 'error', 'error'), una descripción de lo que sucedió, una URL de destino a la que el usuario puede acceder para obtener más información y un "contexto" en caso de que haya varios estados. para una única confirmación. Por ejemplo, un servicio de pruebas puede proporcionar un estado y un servicio de validación como este también puede proporcionar un estado: el campo "contexto" es cómo se diferencian.

Si alguien abre una nueva solicitud de extracción (pull request) en GitHub y este enlace está configurado, es posible que vea algo como Estado de confirmación a través de la API.

### Octokit

Aunque hemos estado haciendo casi todo a través de ```curl``` y simples HTTP solicitudes en estos ejemplos, existen varias bibliotecas de código abierto que hacen que esta API esté disponible de una manera más idiomática. En el momento de escribir este artículo, los lenguajes admitidos incluyen Go, Objective-C, Ruby y .NET. Consulte  https://github.com/octokit para obtener más información sobre estos, ya que manejan gran parte del HTTP por usted.

Con suerte, estas herramientas pueden ayudarlo a personalizar y modificar GitHub para que funcione mejor para sus flujos de trabajo específicos. Para obtener documentación completa sobre toda la API, así como guías para tareas comunes, consulte https://developer.github.com.
